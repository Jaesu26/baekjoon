`-` 왜냐하면 $1$레벨에서 $2$레벨로 가면 힘이 오히려 손해인데 $2$레벨에서 $3$레벨로 가면 힘이 $1$레벨보다 훨씬 오르는 경우를 그리디로는 못 풀기 때문이다

`-` 그리디는 현재 상태에서 최선의 선택만 하기에 $1$레벨에서 $2$레벨로 가지 않을 것이다

`-` 근데 그러면 $1$레벨에서 바로 $3$레벨로 가면 되지 않을까?

`-` 이를 적극 활용하여 문제를 해결해보자

`-` 이제 하루만 남았다고 가정해보자

`-` 그럼 어떤 캐릭터를 레벨 업 시킬 것인가?

`-` 다음 레벨로 갔을 때 힘의 증가량이 가장 큰 캐릭터를 레벨 업 시켜야 가장 이득이다

`-` 그리디로 풀기 위해 $i$레벨에서 $j$레벨로 갔을 때 얻을 수 있는 힘을 계산하자

`-` 이는 for문 $2$번 순회하면 되므로 $O\left(N^2\right)$에 수행할 수 있다

`-` 남은 날짜가 $d$일 때 레벨 업 $i\to j$ 중 남은 날짜 안에 가능한 경우들만 고려하자

`-` $i\to j$에 대해 평균적으로 $1$레벨 업마다 얻는 힘의 증가량을 계산한 뒤 이것이 가장 큰 경우를 선택하여 레벨 업 하자!

`-` 예컨대 레벨 업 $2\to 5$가 $30$의 힘을 얻는다면 레벨 업마다 평균적으로 $10$의 힘을 얻는다

`-` 남은 날짜 $d$ 안에 가능한 경우 중 레벨 업의 평균 힘의 증가량이 가장 큰 $i\to j$를 고르자

`-` $i \to j$로 레벨 업 했을 때 얻는 힘의 증가량을 기록한 뒤 $d$를 $d - (j - i)$로 갱신하자

`-` $d$가 $0$이 될 때까지 반복하면 되고 최종적으로 얻은 힘의 증가량에 초기 보유한 캐릭터의 전체 힘을 합산하면 정답이 된다

`-` 만약 남은 날짜 중 가능한 레벨 업에 대해 레벨 업의 평균 힘의 증가량이 $0$이하인 것들만 남았으면 레벨 업 하는 것이 손해이므로 게임을 하지 않는다

`-` 이러한 방법이 왜 성립하는지 증명하자

`-` 그리디 선택과 정답 선택에 대해 동일한 레벨 업 경우는 제외하고 남은 것들을 고려할 것이다

`-` 그리디 선택으로 $i \to j$ 레벨 업을 골랐다 해보자 (정답 선택에서는 고르지 않았다)

`-` $i\to j$는 레벨 업당 평균 $a$의 힘을 얻는다

`-` 정답 선택에선 $j-i$ 날짜 동안 다르게 레벨 업 했다

`-` 그런데 해당 레벨 업 모두 레벨 업당 평균 $a$의 힘보다 적은 힘을 얻는다

`-` 따라서 해당 선택을 $i\to j$로 변경하면 더 큰 힘을 얻을 수 있으므로 그리디 선택이 정답 선택보다 이득이다

`-` 따라서 귀류법에 의해 그리디 선택이 올바르다

`-` 전체 알고리즘의 시간 복잡도는 최악의 경우 $O\left(M N^2\right)$이다

`-` 테스트케이스는 다 맞혔는데 $19\%$에서 틀렸다

`-` 왜 그런지 생각해보니까 캐릭터가 별로 없으면 레벨 업 순서가 중요할 때가 존재한다

`-` $1\to5$로 가면 $210$을 얻고 $ 5\to6$이면 $100$을 얻는다고 해보자

`-` 평균으로 따지면 $5 \to 6$이 맞다

`-` 하지만 $5$렙 캐릭터가 하나뿐이고 날짜가 얼마 안 남았다면?

`-` $5 \to 6$을 포기하고 $1 \to 5$를 선택해서 $100$ 대신 $210$을 획득해야 한다

```python
def compute_gain_by_level_up(level2power):
    level_up2gain = {(0, 0): 0}
    for i in range(1, N):
        for j in range(i + 1, N + 1):
            level_up2gain[(i, j)] = level2power[j] - level2power[i]
    return level_up2gain


def find_best_level_up(level2num, level_up2gain, day):
    max_avg_gain_per_level = 0
    level_up = 0, 0  # i -> j
    for (i, j), gain in level_up2gain.items():
        if j - i > day or level2num[i] == 0:
            continue
        avg_gain_per_level = gain / (j - i)
        if avg_gain_per_level <= max_avg_gain_per_level and :
            continue
        max_avg_gain_per_level = avg_gain_per_level
        level_up = i, j
    return level_up


def level_up_greedy(level2num, level_up2gain, day):
    total_gain = 0
    while day > 0:
        level_before, level_after = find_best_level_up(level2num, level_up2gain, day)
        gain = level_up2gain[(level_before, level_after)]
        if gain <= 0:
            break
        total_gain += gain
        level2num[level_before] -= 1
        level2num[level_after] += 1
        day -= level_after - level_before
    return total_gain

    
def solution():
    global N
    N = int(input())
    level2num = [0] + [*map(int, input().split())]
    level2power = [0] + [*map(int, input().split())]
    D = int(input())
    level_up2gain = compute_gain_by_level_up(level2power)
    default_power = sum(level2num[i] * level2power[i] for i in range(1, N + 1))
    answer = default_power + level_up_greedy(level2num, level_up2gain, D)
    print(answer)


solution()
```

### 정답 풀이 (동적 계획법)

`-` 그리디로 풀지 못한다는 것을 알았지만 대신 동적 계획법 접근 방식이 떠올랐다

`-` 일단 어떤 레벨의 캐릭터의 수가 $D$를 초과하는 건 의미가 없다

`-` 어차피 $D$일 동안 레벨 업을 최대 $D$번 할 수 있기 때문에 그 이상의 캐릭터는 사용하지 않기 때문이다

`-` 또한 레벨이 높아질 수는 있어도 낮아질 수는 없다

`-` $1$레벨 캐릭의 수를 $x_1$이라 하자

`-` 그럼 정답을 분석하여 $1$레벨 캐릭터를 몇 명 훈련시켰는지 따져보면 $1$레벨 캐릭터 중 $1,2,\dots, \min(x_1, D)$명을 훈련시킨 것 중 하나이다

`-` 그렇게 $1$레벨 캐릭터를 훈련시켜 레벨 업 시키면 $2$가 되고 레벨은 더 낮아지지 않으므로 앞으로 $1$레벨 캐릭터를 훈련시키는 것을 고려하지 않아도 된다

`-` 그럼 현재 고려하고 있는 레벨의 캐릭터 수와 남은 일수를 알고 있을 때 현재까지 얻은 힘 중 최댓값을 알 수 있다

`-` $1$레벨 캐릭터를 $i$명 훈련시켰다고 해보자

`-` 그럼 $2$레벨 캐릭터는 $x_2 + i$명이 되고 그 때의 남은 일수는 $D - i$이며 얻은 힘은 $f(1, 2) \times i$이다 ($f(i,j)$는 $i$에서 $j$로 레벨 업 했을 때 얻은 힘)

`-` 현재 고려하고 있는 레벨의 캐릭터 수가 동일하고($D$ 이상이어도 동일) 남은 일수도 동일하다면 얻은 힘이 커야 이득이다

`-` 각 레벨의 상태 공간의 크기는 $O\left(D^2\right)$이며 레벨은 $N$개 존재한다

`-` 따라서 전체 상태 공간의 크기는 $O\left(ND^2\right)$이고 현재 레벨에서 다음 레벨로 갱신하는 건 $O(D)$이다

`-` 따라서 $O\left(ND^3\right)$의 시간 복잡도로 문제를 해결할 수 있다

`-` $12\%$에서 틀려서 디버깅 했는데 너무 오래 걸렸다

`-` dp를 배열로 관리하여 캐릭터 수가 $D$를 초과하면 커팅하는 식으로 했는데 딕셔너리로 바꾸자마자 정답을 맞혔다

`-` $4.5$시간이나 걸렸다...

````python
def compute_gain_by_level_up(level2power):
    level_up2gain = {}
    for i in range(1, N):
        for j in range(i + 1, N + 1):
            level_up2gain[(i, j)] = level2power[j] - level2power[i]
    return level_up2gain


def level_up_dp(level2num, level_up2gain):
    dp = {}  # dp[(캐릭터 수, 남은 일수)] = 현재까지 얻은 힘
    dp[(level2num[1], D)] = 0
    for level in range(1, N):
        gain = level_up2gain[(level, level + 1)]
        dp_next = {}
        for (n, d), power in dp.items():
            for num_trainings in range(min(n, d) + 1):
                t = num_trainings 
                if power + t * gain <= dp_next.get((level2num[level + 1] + t, d - t), -INF):
                    continue
                dp_next[(level2num[level + 1] + t, d - t)] = power + t * gain
        dp = dp_next
    return max(dp.values())


def solution():
    global N, D, INF
    N = int(input())
    level2num = [0] + [*map(int, input().split())]
    level2power = [0] + [*map(int, input().split())]
    D = int(input())
    INF = float("inf")
    level_up2gain = compute_gain_by_level_up(level2power)
    default_power = sum(level2num[i] * level2power[i] for i in range(1, N + 1))
    level2num = [min(n, D) for n in level2num]
    answer = default_power + level_up_dp(level2num, level_up2gain)
    print(answer)


solution()

# input
# 5
# 1 2 3 4 5
# 1 4 9 16 25
# 10
````

```python
import copy


def check_arrangement(grid, row, col, piece):
    n, m = len(piece), len(piece[0])
    for r in range(row, row + n):
        for c in range(col, col + m):
            if piece[r - row][c - col] == PIECE and grid[r][c] != BLANK:
                return False
    return True


def arrange_piece(grid, row, col, piece, id_):
    n, m = len(piece), len(piece[0])
    for r in range(row, row + n):
        for c in range(col, col + m):
            if piece[r - row][c - col] != PIECE:
                continue
            grid[r][c] = id_


def rollback_grid(grid, row, col, piece):
    n, m = len(piece), len(piece[0])
    for r in range(row, row + n):
        for c in range(col, col + m):
            if piece[r - row][c - col] != PIECE:
                continue
            grid[r][c] = BLANK


def backtracking(grid, pieces, id_):
    global COMPLETE, RESTORED
    if COMPLETE:
        return
    if id_ > 5:
        COMPLETE = True
        RESTORED = copy.deepcopy(grid)
        return
    piece = pieces[id_]
    n, m = len(piece), len(piece[0])
    for row in range(L):
        for col in range(L):
            if row + n - 1 >= L or col + m - 1 >= L:
                continue
            if not check_arrangement(grid, row, col, piece):
                continue
            arrange_piece(grid, row, col, piece, id_)
            backtracking(grid, pieces, id_ + 1)
            rollback_grid(grid, row, col, piece)


def solution():
    global L, BLANK, PIECE, COMPLETE, RESTORED
    L = int(input())
    PIECE = "#"
    BLANK = "."
    COMPLETE = False
    RESTORED = None
    pieces = {}
    area = 0
    for i in range(1, 6):
        N, M = map(int, input().split())
        pieces[i] = [input() for _ in range(N)]
        area += sum(row.count(PIECE) for row in pieces[i])
    if area != L**2:
        print("gg")
        return
    grid = [[BLANK] * L for _ in range(L)]
    backtracking(grid, pieces, 1)
    if not COMPLETE:
        print("gg")
        return
```

### 정답 풀이 (이분 탐색)

`-` 처음엔 단순하게 접근해보자

`-` $i$가 고정됐다면 $T_i = \max(\{D_{i1}, \dots D_{in}\})$을 계산하는데 $O(N)$이고 $i$는 $1$부터 $N$까지 가능하다

`-` 따라서 나이브한 풀이의 시간 복잡도는 $O\left(N^2\right)$이다

`-` $N$은 최대 $300000$이므로 $O\left(N^2\right)$ 풀이는 시간 초과이다

`-` $D_{ij}$에서 $|i - j|$항 때문에 동일한 $j_1, j_2$에 대해 $i$에 따라 $D_{ij_1}, D_{ij_2}$의 대소관계가 한 번 역전될 수 있다

`-` 예컨대 $a_{5} = 15, a_{6} = 10$이라고 해보자

`-` $D_{4,5} = 15, D_{4,6} = 20$이므로 $D_{4,5} < D_{4,6}$이지만 $D_{2, 5} = 45, D_{2, 6} = 40$이므로 $D_{2,5} > D_{2,6}$이다

`-` 최대 한 번 역전될 수 있는 것을 어떻게 처리하면 좋을지 고민하는데 시간을 많이 썼다

`-` $T_x$가 문제의 정답인 $x$를 생각해보자

`-` 이러한 $x$는 여러 개 있을 수 있는데 어차피 $T$는 동일하테니 뭐든 상관 없다

`-` $x$ 위치에 히터를 설치했다고 해보자

`-` 히터를 오른쪽으로 옮기거나 왼쪽으로 옮기거나 할 수 있을 것이다

`-` 그런데 어느 쪽으로 옮기든 기존의 $T_x$보다 커질 수 밖에 없다

`-` 만약 그렇지 않다면 $T_x$가 최소라는 가정에 모순되기 때문이다

`-` 다르게 말하면 어떤 임의의 위치 $i$에 대해 $i$의 오른쪽으로 가든 왼쪽으로 가든 시간 $T$가 줄지 않는다면 $i$가 최적의 위치인 것이다

`-` 현재 위치와 정답 위치에 따라 가능한 상황을 정리해보자

`-` 현재 위치를 $i$라 하고 정답 위치를 $x$라 하겠다

`-` $i = x$라면 왼쪽으로 움직이든 오른쪽으로 움직이든 시간이 줄지 않는다

`-` $i > x$라면 정답 위치보다 오른쪽에 있다는 뜻이다

`-` 여기서 왼쪽으로 움직여야 시간이 줄어들게 된다

`-` 이는 $i$를 기준으로 하여 $D_{ij}$를 계산했을 때 $D_{ij}$가 최대일 때의 $j$가 $i$보다 작다는 의미이다

`-` 이를 줄이기 위해 왼쪽으로 움직여야 하는 것이다

`-` $i < x$인 경우는 $i > x$와 대칭적이니 설명은 생략한다

`-` 움직일 때마다 현재의 $i$에 대해 $T_i$를 계산한 후 여태까지 고려한 $T$들의 최솟값인 $T_{\min}$보다 작다면 갱신한다

`-` 만약 새롭게 $j$로 움직였는데 $T_j$가 $T_{\min}$보다 작지 않다면 어디로 움직여도 시간이 줄지 않는다는 것이고 이는 $T_{\min}$에 대해 $i=x$인 상황이다 (사실 아니었음...)

`-` 단순히 한 칸 씩 움직이는 건 비효율적이므로 이분 탐색을 사용할 것이다

`-` 현재 위치 $i$와 정답 위치 $x$에 따라 움직여야 하는 방향은 고정되어 있고 얼만큼 움직이냐만 고려하면 된다

`-` 방향이 고정되어 있으니 이분 탐색을 활용해 움직일 거리를 정하자

`-` 탐색을 $O(\log N)$번 수행하고 각 탐색마다 $T$를 계산하기 위해 $O(N)$의 비용을 소모하므로 전체 알고리즘의 시간 복잡도는 $O(N\log N)$이다

`-` 틀려서 봤더니 약간의 문제가 있었다

`-` 새롭게 움직였는데 $T_{\min}$을 갱신하지 못하면 탐색을 멈췄다

`-` 그런데 이것은 잘못됐다

`-` 예컨대 현재의 바로 오른쪽으로 위치를 옮기는 것이 정답이여도 이분 탐색에선 탐색할 영역의 중간 지점으로 움직인다

`-` 이때 $T_{\min}$을 갱신하지 못 할 수 있지만 탐색을 계속하면 결국엔 정답 위치를 찾는다

`-` $T_{\min}$을 갱신하지 못해도 탐색을 계속하도록 수정하니까 맞혔다

`-` $1$시간 $40$분 걸렸다

`-` 소신발언: 풀이 작성하는게 제일 어려움 ($1$시간 $40$분 중 풀이 작성하는데 $1$시간 썼을 듯;)

````python
def argindex_heater_operating_time(array, i):
    d_max = 0
    d_argmax = i
    a_i = array[i]
    for j in range(N):
        a_j = array[j]
        d_ij = abs(i - j) * a_j
        if d_ij > d_max:
            d_max = d_ij
            d_argmax = j
    return d_argmax


def binary_search(array, start, end):
    t_min = INF
    while start <= end:
        mid = (start + end) // 2
        t_argindex = argindex_heater_operating_time(array, mid)
        t = abs(mid - t_argindex ) * array[t_argindex]
        if t < t_min:
            t_min = t
        if t_argindex <= mid:
            end = mid - 1
        else:
            start = mid + 1
    return t_min


def solution():
    global N, INF
    N = int(input())
    array = [*map(int, input().split())]
    INF = float("inf")
    answer = binary_search(array, 0, N - 1)
    print(answer)


solution()

# input
# 3
# 100 4 3
````

`-` 따라서 굳이 모든 상자의 좌표를 순회할 필요가 없다

`-` 예컨대 자동차가 현재 $(3, 5)$에 있고 $x$좌표가 증가하는 방향으로 $1000$만큼 움직인다고 해보자

`-` 그럼 자동차는 $(1003, 5)$에 도착하게 되고 지나온 좌표에 있는 상자를 모두 획득할 수 있다

`-` $y=5$가 아닌 상자는 확인할 필요가 없는 것이다

`-` 이를 위해 $i$번 상자가 입력으로 들어오면 $x$ 좌표인 $x_i$와 $y$ 좌표인 $y_i$에 $i$번 상자가 있음을 표기할 것이다

`-` 이는 딕셔너리를 사용할 것이고 $d_x[x_i]$에 $y_i$를 추가하고 $d_y[y_i]$에는 $x_i$를 추가하여 구현할 것이다

`-` 모든 상자들의 좌표는 다르므로 중복될 일은 없다

`-` 위의 예시를 가지고 설명하면 $d_y[5]$를 순회하여 값이 $4$(출발지는 포함 안 함)와 $1000$ 사이에 있으면 해당 좌표에 있는 상자의 아이템을 얻을 수 있는 것이다

`-` 상자의 들어있는 아이템 개수는 상자의 좌표에 아이템 개수를 맵핑한 딕셔너리로 관리할 것이다

`-` 그런데 $d_y[5]$에 존재하는 상자가 최악의 경우 $N$개이므로 쿼리 하나당 $O(N)$인데 쿼리가 $Q$개 이므로 총 $O(NQ)$이다

`-` $1 \le Q \le 200000$이므로 위의 방식은 시간 초과이다

`-` 굳이 $d_y[5]$를 모두 순회할 필요없이 구간 합과 이분 탐색을 이용해 시간을 줄일 것이다

`-` 일단 $d_y[5]$를 정렬해두자

`-` 그럼 이분 탐색을 활용해 임의의 값을 $O(\log N)$에 찾을 수 있다

`-` 예를 들어 자동차가 $(3, 5)$에서 $(1003, 5)$까지 움직여 $d_y[5]$에서 출발 좌표 $3$과 도착 좌표 $1003$을 찾는다고 해보자

`-` $d_y[5]$에 $3$이 없을 수도 있다

`-` 없으면 어차피 상자를 못 얻으니 상관없고 $3$보다 큰 값을 찾으면 된다 (출발 좌표에 있는 상자는 못 줍는다)

`-` $3$보다 큰 값의 인덱스를 $i$라 하고 $1003$보다 작거나 같은 값의 인덱스를 $j$라 하자

`-` 그럼 자동차가 움직여서 $j - i + 1$개의 상자를 주운 것이고 이를 $O(\log N)$에 계산했다

`-` 그런데 이것만 알면 되는 것은 아니고 상자에 들어있는 아이템의 개수를 알아야 한다

`-` 당연히 인덱스 $i$부터 $j$까지 순회하며 각 상자에 들어있는 아이템의 개수를 취합하는 건 $O(N)$이므로 시간 초과이다

`-` 이를 빠르게 계산하기 위해 구간 합을 사용할 것이다

`-` $p_{y=5}[i]$를 $y=5$인 상자들을 $x$ 좌표를 기준으로 정렬했을 때 첫 번째 상자부터 $i$번째 상자까지 주워서 얻은 아이템의 총 개수라고 하자

`-` 이는 구간 합 테크닉을 통해 $O(N)$에 계산할 수 있다

`-` 하나의 상자에 대해 $x,y$ 좌표로 나눠서 $2$번만 중복해 계산하므로 $d_x, d_y$ 별로 각 상자에 대해 구간 합을 계산해도 $O(N)$이다

`-` 아무튼 $i$번 상자부터 $j$번 상자까지 주워서 얻은 아이템의 총 개수는 $p_{y=5}[j] - p_{y=5}[i - 1]$이다

`-` 좌표 정렬은 $O(N \log N)$이고 구간 합을 사전 계산 하는 건 $O(N)$이며 쿼리마다 얻은 아이템을 계산하는 건 $O(\log N)$이다

`-` 쿼리는 $Q$번 주어지므로 전체 알고리즘의 시간 복잡도는 $O\left((Q + N)\log N\right)$이다

`-` 쿼리마다 얻은 아이템의 개수를 출력하는게 아니고 전체 쿼리가 끝났을 때 얻은 아이템의 개수를 출력해야 된다

`-` $2$시간 걸렸다 ($1$트 컷 ㄷㄷ;)

`-` 구현이 힘들었다

`-` 코드 더러워서 리팩토링 좀 했다

````python
from collections import defaultdict


def compute_prefix_sum(boxes):
    n = len(boxes)
    prefix_sum = [0] * (n + 1) 
    for i, (_, w) in enumerate(boxes, start=1):
        prefix_sum[i] = prefix_sum[i - 1] + w
    return prefix_sum


def preprocess(axis2boxes):
    axis2prefix_sum = {}
    for axis in axis2boxes:
        axis2boxes[axis].sort()
        prefix_sum = compute_prefix_sum(axis2boxes[axis])
        axis2prefix_sum[axis] = prefix_sum
    return axis2prefix_sum


def binary_search(target, boxes, prefix_sum):
    left = 0
    right = len(boxes) - 1
    while left <= right:
        mid = (left + right) // 2
        if boxes[mid][AXIS] <= target:
            left = mid + 1
        else:
            right = mid - 1
    return left, right


def search_argmin_among_over(target, boxes, prefix_sum):
    if boxes[-1][AXIS] <= target:
        return NONE
    left, _ = binary_search(target, boxes, prefix_sum)
    return left


def search_argmax_among_at_most(target, boxes, prefix_sum):
    if boxes[0][AXIS] > target:
        return NONE
    _, right = binary_search(target, boxes, prefix_sum)
    return right


def forward_acquire_items_along_axis(start, end, boxes, prefix_sum):
    i = search_argmin_among_over(start, boxes, prefix_sum)
    j = search_argmax_among_at_most(end, boxes, prefix_sum)
    if i == NONE or j == NONE:
        return 0
    return prefix_sum[j + 1] - prefix_sum[i]


def search_argmax_among_under(target, boxes, prefix_sum):
    if boxes[0][AXIS] >= target:
        return NONE
    if boxes[-1][AXIS] < target:
        return len(boxes) - 1
    _, right = binary_search(target, boxes, prefix_sum)
    if boxes[right][AXIS] == target:
        return right - 1
    return right


def search_argmin_among_at_least(target, boxes, prefix_sum):
    if boxes[-1][AXIS] < target:
        return NONE
    if boxes[0][AXIS] >= target:
        return 0
    left, _ = binary_search(target, boxes, prefix_sum)
    if boxes[left - 1][AXIS] == target:
        return left - 1
    return left


def backward_acquire_items_along_axis(start, end, boxes, prefix_sum):
    i = search_argmin_among_at_least(end, boxes, prefix_sum)
    j = search_argmax_among_under(start, boxes, prefix_sum)
    if i == NONE or j == NONE:
        return 0
    return prefix_sum[j + 1] - prefix_sum[i]


def query(d, v, x, y, x2ys, y2xs, x2prefix_sum, y2prefix_sum):
    num_acquired_items = 0
    if d == 0:
        if y in y2xs:
            num_acquired_items = forward_acquire_items_along_axis(x, x + v, y2xs[y], y2prefix_sum[y])
        x += v
    elif d == 1:
        if x in x2ys:
            num_acquired_items = forward_acquire_items_along_axis(y, y + v, x2ys[x], x2prefix_sum[x])
        y += v
    elif d == 2:
        if y in y2xs:
            num_acquired_items = backward_acquire_items_along_axis(x, x - v, y2xs[y], y2prefix_sum[y])
        x -= v
    elif d == 3:
        if x in x2ys:
            num_acquired_items = backward_acquire_items_along_axis(y, y - v, x2ys[x], x2prefix_sum[x])
        y -= v
    return x, y, num_acquired_items


def solution():
    global AXIS, NONE                      
    N, Q = map(int, input().split())
    AXIS = 0
    NONE = -1
    x2ys = defaultdict(list)
    y2xs = defaultdict(list)
    for _ in range(N):
        x, y, w = map(int, input().split())
        x2ys[x].append((y, w))
        y2xs[y].append((x, w))
    x2prefix_sum = preprocess(x2ys)
    y2prefix_sum = preprocess(y2xs)
    x, y = 1, 1
    total_acquired_items = 0
    for _ in range(Q):
        d, v = map(int, input().split())
        x, y, num_acquired_items = query(d, v, x, y, x2ys, y2xs, x2prefix_sum, y2prefix_sum)
        total_acquired_items += num_acquired_items
    print(total_acquired_items)


solution()

# input
# 3 3
# 1 3 1
# 2 2 1
# 3 1 1
# 1 3
# 0 2
# 3 3
````

### 시간 초과 풀이 (브루트 포스)

`-` 가장 큰 책이 맨 아래에 있다면 해당 책은 더 이상 정렬에 영향을 주지 않는다 (하노이의 탑과 비슷함)

`-` 일단 최악의 경우 정렬하는데 몇 단계가 필요한지 알아야 한다

`-` 근데 정확히 몇 단계가 필요한지 잘 모르겠다

`-` 예제를 보면 $n=8$일 때 하한이 $53$, $n=22$일 때 하한이 $1234567$이다

`-` 매 단계마다 책을 빼서 위의 놓는 건 $O\left(n\right)$이다

`-` $f(n)$을 책이 $n$권일 때 정렬하는데 필요한 단계 수의 최댓값이라 하면 총 시간 복잡도는 최악의 경우 $O\left(n f(n)\right)$이다

`-` $n\le 50$이므로 $f(n)$이 크지 않으면 브루트 포스로 풀릴 것 같다

`-` 매 단계마다 정렬되지 않았다면 책을 빼서 가장 위에 놓아야 한다

`-` 따라서 스택의 가장 마지막 원소가 책의 가장 위에 놓이도록 조정해 추가 연산을 $O(1)$에 수행하겠다

`-` 단순한 풀이는 시간 초과였다

```python
def step(stack):

        if stack[i] > stack[i - 1]:
```

### 정답 풀이 (동적 계획법)

`-` 위에서부터 정렬된 스택 $s_1, s_2, \dots, s_i$를 고려하자

`-` 정렬된 스택 아래에 크기가 $x$인 새로운 책을 추가한다고 해보자

`-` $x$를 추가하여 새롭게 정렬된 스택이 $s_1, s_2, \dots, s_{k-1}, x, s_k, \dots, s_i$라고 하자

`-` $x \le s_k$이므로 $s_k$ 아래의 책은 $x, s_1, s_2, \dots, s_{k-1}$를 정렬하는데 영향을 끼치지 않는다

`-` 즉, $x, s_1, s_2, \dots, s_{k-1}$를 정렬하는데 필요한 단계 수와 $x, s_1, s_2, \dots, s_i$를 정렬하는데 필요한 단계 수가 동일하다

`-` 따라서 $x$보다 작은 부분에 한해서 정렬에 필요한 단계 수를 계산하면 된다

`-` 책의 크기가 모두 다르다면 정렬하는데 몇 단계 필요한지 계산할 수 있다

`-` 서로 다른 크기를 가진 정렬된 스택 $s_1, s_2, \dots, s_i$와 가장 큰 책 $x$를 고려하자

`-` $g(n)$을 서로 다른 크기의 정렬된 책 $n$권이 존재하고 맨 위에 가장 큰 책을 올렸을 때 이를 정렬하는데 필요한 최소 단계 수라고 하자

`-` 그럼 $x, s_1, s_2, \dots, s_i$를 정렬하는데 필요한 단계 수는 $g(i)$이다

`-` $s_i$를 빼서 맨 위에 놓으려면 $s_i$ 위의 책들이 정렬되어 안정된 상태여야 한다

`-` 즉, $s_1, s_2, \dots, x, s_i$인 상태를 먼저 만들어야 한다

`-` 이건 $x, s_1, s_2, \dots, s_{i-1}$을 정렬한 뒤 $s_i$를 추가한 것과 같다

`-` $x, s_1, s_2, \dots, s_{i-1}$를 정렬하는데 필요한 단계 수는 $g(i-1)$이다

`-` $s_1, s_2, \dots, x, s_i$인 상태에서 $s_i$를 뺀 뒤 맨 위의 올리는 건 $1$단계가 필요하며 상태는 다음처럼 변한다

`-` $s_i, s_1, s_2, \dots, s_{i-1}, x$에서 $x$는 가장 크며 맨 아래에 있으므로 $s_i, s_1, s_2, \dots, s_{i-1}$을 정렬하는데 영향을 주지 않는다

`-` $s_i, s_1, s_2, \dots, s_{i-1}$을 정렬하는데 필요한 단계 수는 $g(i-1)$이다

`-` 따라서 $g(i) = 2g(i-1) + 1$이고 $g(0)=0, g(1)=1$이므로 $g(i) = 2^i - 1$이다

`-` 이는 책의 크기가 모두 다를 때 성립하고 크기가 동일한 책이 존재하면 더 작은 값을 가지게 된다

`-` 아무튼 원래로 돌아가서 $x, s_1, s_2, \dots, s_{k-1}$를 정렬하는 걸 고려하고 정렬에 필요한 최소 단계 수를 $g'(k-1)$이라 하자

`-` 참고로 함수 $g'$은 $g$와 달리 스택의 원소에 영향을 받는다

`-` 일단 $s_{k-1} < x$이므로 정렬된 스택은 $s_1, s_2, \dots, s_{k-1}, x$이다

`-` 이제 $s_{k-1}$이 유일하게 가장 큰 경우와 그렇지 않은 경우로 나눌 수 있다

`-` $s_{k-1}$이 유일하게 크다고 해보자

`-` 그럼 $x, s_1, s_2, \dots, s_{k-2}$를 정렬하는데 $g'(k-2)$의 단계가 필요하고 결과는 $ s_1, s_2, \dots, s_{k-2}, x$가 된다

`-` 이제 $1$단계를 소모해 $s_{k-1}$을 맨 위로 올린 뒤 $s_{k-1}, s_1, s_2, \dots, s_{k-2}, x$을 정렬하면 된다

`-` 이때 $x$가 맨 아래 있고 가장 크므로 실질적으로 $s_{k-1}, s_1, s_2, \dots, s_{k-2}$를 정렬하면 되고 필요한 단계 수는 $g'(k-2)$이다

`-` 따라서 정렬에 필요한 총 단계 수는 $2g'(k-2) + 1$이다

`-` 이제 다른 경우로 $s_{k-1}$이 유일하게 크지 않고 중복된 값이 존재한다고 해보자

`-` 그럼 $x, s_1, s_2, \dots, s_{k-1}$에서 임의의 구간 $s_j$부터 $s_{k-2}$까지 $s_{k-1}$과 동일하다

`-` 우선 $g'(j-1)$단계를 거쳐 $s_1, s_2, \dots, s_{j-1}, x, s_j, \dots, s_{k-1}$로 정렬하자

`-` $s_j$를 맨 위로 올리는데 $1$단계가 필요하고 정렬하는데 $g'(j-1)$단계가 필요하다

`-` 그럼 $s_1, s_2, \dots, s_{j-1}, s_j, x, s_{j+1}, \dots, s_{k-1}$이 된다

`-` 이를 $s_{k-1}$까지 반복하면 되고 $x, s_1, s_2, \dots, s_{k-1}$를 정렬하는데 총 $(h + 1)g'(j-1) +h$단계가 필요하다 ($h=k-j$)

`-` 점화식이 주어졌고 기본 값으로 $g'(0)=0, g'(1)=1$이므로 중복 크기를 포함하는 스택도 정렬하는데 필요한 단계 수를 계산할 수 있다

`-` $i$를 $1$부터 $n$까지 키우며 스택의 맨 위부터 $i$번째 구간까지 정렬해 나가면 된다

`-` 정렬마다 필요한 단계 수를 계산하는 건 $O(n)$이 소요되지만 실제 정렬엔 $O(n\log n)$이 소요된다

`-` 스택의 $1\sim i$번째 구간을 정렬하는데 $i$가 $1$부터 $n$까지 변하므로 총 $n$번 정렬해야 한다

`-` 따라서 총 알고리즘의 시간 복잡도는 $O\left(n^2\log n\right)$이다

````python
from collections import Counter


def count_step(stack, x):
    if not stack or stack[-1] <= x:
        return 0
    for i, s in enumerate(stack):
        if x <= stack[i]:
            index = i
            break
    part = stack[:index].copy()
    part.insert(0, x)
    return g(part) + 1


def g(stack):
    if len(stack) <= 1:
        return 0
    s_k = stack[-1]
    h = Counter(stack)[s_k]
    if h == 1:
        return 2 * g(stack[:-1]) + 1
    return (h + 1) * g(stack[:-h]) + h


def solve_testcase():
    n = int(input())
    stack = list(map(int, input().split()))
    answer = 0
    for i in range(n):
        part = sorted(stack[:i])
        x = stack[i]
        answer += count_step(part, x)
    print(answer)


def solution():
    T = int(input())
    for _ in range(T):
        solve_testcase()


solution()

# input
# 4
# 4
# 3 4 1 2
# 8
# 3 1 4 1 5 9 2 6
# 5
# 1 42 42 42 1000
# 22
# 4 1 2 5 6 7 9 10 3 13 17 11 12 14 19 20 22 8 15 16 18 21
````

`-` 근데 중간에 배 아파서 집중 못 하고 화장실도 갔다 옴

### 정답 풀이 (애드 혹, 많은 조건 분기)

`-` $N\le10^6$이므로 최소 $O\left(N\log N\right)$에는 해결해야 하니 쉽게 생각하려고 $1,2,3$의 개수에 따라 접근해봤다

`-` $1,2,3$의 개수에 따라 몇 번의 변경이 필요한지 나누던 중 배열의 양 끝 원소가 중요하다는 것을 깨달았다

`-` 일단 배열의 크기가 $1$인데 원소가 $3$이라면 헤어 스타일을 세갈래가 아니게 만들 수 없으므로 정답은 $-1$이다

`-` 모든 사람의 헤어 스타일이 세갈래라고 해보자

`-` 사람의 수가 짝수라면 전체를 xor해서 $3$이 아니게 만들 수 있고 홀수라면 $3$번의 변경이 필요하다

`-` 만약 모든 사람의 헤어 스타일이 생머리 또는 세갈래라고 해보자

`-` 세갈래의 수가 짝수라면 전체를 xor해서 $3$이 아니게 만들 수 있다

`-` 세갈래의 수가 홀수라면 $0$과 $3$을 xor해서 $3$의 개수를 짝수로 만든 뒤 전체를 xor하면 모든 사람의 머리를 생머리로 만들 수 있다

`-` 이제 포니테일($1$) 또는 양갈래($2$)의 머리를 가진 사람들이 존재한다고 해보자

`-` 간단하게 전체를 xor해서 $3$이 아니라면 정답은 $1$이 된다

`-` 그렇지 않다면 배열에서 $0$을 제거하자 ($0$과 자기 자신의 xor 결과는 자기 자신이므로 $0$은 있어도 의미가 없다) 

`-` 배열의 양 끝 원소 중 하나가 $1$ 또는 $2$라고 해보자

`-` 전체를 xor했을 때 $3$이므로 $1$ 또는 $2$인 끝의 원소를 제외하고 xor을 수행하면 $3$이 아니게 된다

`-` 따라서 $1$번의 변경으로 세갈래인 사람이 아무도 없게 만들 수 있다

`-` 마지막으로 배열의 양 끝 원소 모두 $3$이라고 해보자

`-` 왼쪽 끝에서 처음으로 $1$ 또는 $2$가 나올 때까지 중앙으로 구간을 넓히자

`-` 해당 구간을 xor하면 $1$ 또는 $2$가 $1$개뿐이므로 결괏값은 $3$이 아니게 된다

`-` xor 연산은 결합법칙이 성립하므로 전체에서 해당 구간을 제외한 나머지 구간끼리 xor해도 $3$이 아니게 된다 ($3$이면 모순)

`-` 따라서 $2$번의 변경으로 세갈래인 사람이 아무도 없게 할 수 있다

`-` $0$이 무쓸모일 줄 알아서 계속 틀렸었는데 배열에 $0,3$만 존재하면 쓸모가 있는 걸 깨닳고 정답을 맞혔다

`-` 알고리즘의 시간 복잡도는 $O\left(N\right)$이다

`-` $2$시간 $20$분 걸렸다

`-` 원래 안 쉬고 끝까지 풀어야 되는데 중간에 너무 졸려서 자고 왔다 (자는 시간은 풀이 시간에서 제외함)

````python
def xor(array):
    result = 0
    for a in array:
        result ^= a
    return result


def make_non_triple_ponytails(hair_styles):
    uniques = set(hair_styles)
    if 3 not in uniques:
        return 0
    result = xor(hair_styles)
    if result != 3:
        return 1
    if len(uniques) == 1:  # Only 3
        if len(hair_styles) == 1:
            return -1
        return 3
    if len(uniques) == 2 and 0 in uniques:
        return 2
    hair_styles = [h for h in hair_styles if h != 0]
    if hair_styles[0] != 3 or hair_styles[-1] != 3:
        return 1
    return 2


def solve_testcase():
    N = int(input())
    hair_styles = [*map(int, input().split())]
    answer = make_non_triple_ponytails(hair_styles)
    print(answer)


def solution():
    T = int(input())
    for _ in range(T):
        solve_testcase()


solution()

# input
# 5
# 4
# 3 0 1 2
# 4
# 1 3 3 3
# 4
# 3 1 2 3
# 3
# 3 3 3
# 1
# 3
````

`-` xor 연산을 eval 함수를 이용하면 RecursionError가 발생한다 (신기함)

`-` 즉, 입력을 문자열로 받고 xor 연산을 `eval("^".join(hair_styles))`로 수행하면 RecursionError가 발생한다

`-` 재귀 깊이를 백만으로 설정해도 PyPy가 아닌 Python으로 제출하면 오류가 발생한다

`-` 재귀 깊이를 $10^7$로 설정해도 Python에선 RecursionError가 발생한다

`-` 그리고 eval 함수는 너무 느리다

`-` eval 함수 안 사용하니까 시간이 $\frac{1}{15}$로 단축돼서 $168\operatorname{ms}$로 $1$등 됐다

## 상근이의 레스토랑

- 문제 출처: [백준 2786번](https://www.acmicpc.net/problem/2786)

### 정답 풀이 (동적 계획법, 정렬)

`-` 일단 $1$개를 주문할 땐 $A_i$가 가장 작은 음식을 주문하면 된다

`-` 여태까지 $Q$개의 음식을 최소 금액으로 주문했다고 해보자

`-` $Q$개의 음식을 최소 금액으로 주문했을 때의 비용을 $X_Q$라 하자

`-` $Q+1$개의 음식을 주문할 때 처음에 주문하는 음식은 $Q$개의 음식을 주문했을 때 처음에 주문한 음식이거나 아니다

`-` 만약 $Q$개의 음식을 주문했을 때 처음에 주문한 음식을 그대로 주문한다면 단순히 남은 음식 중 $B_i$가 가장 작은 음식을 시키면 된다

`-` 그럼 $X_{Q+1} = X_Q + B_i$이다

`-` 그렇지 않다고 해보자

`-` 처음 주문한 음식의 가격은 $A_j, B_j$이고 새롭게 처음으로 주문할 음식의 가격은 $A_k, B_k$라고 하자

`-` 그럼 $X_{Q+1} = X_Q + A_k - A_j + B_j$이다

`-` $B_i$와 $A_k - A_j + B_j$중에서 더 작은 값의 음식을 주문하면 된다

`-` 이때 $- A_j + B_j$는 $k$와 관계 없이 고정된 상수이므로 $A_k$가 가장 작은 음식을 선택해서 비교하면 된다 ($A_k - B_k$가 기준이 아님)

`-` 즉, $X_{Q+1} = X_Q + \min(B_i, A_k - A_j + B_j)$, (단, $A_j,B_j$는 $X_Q$를 도출하는 주문 순서에서 처음 주문한 음식의 가격)

`-` 음식을 $1,2,\dots$개 구매하는 매 단계에서 $A_j,B_j$는 고정된 상수이다

`-` 즉, $X_{Q+1}$을 고려할 때 $X_Q$에서의 처음 주문한 음식이 $j$가 아닌 다른 $s$였을 때의 도출된 금액 $X^{s}_Q$를 고려하지 않는다 ($j,s$는 $Q$개에 속해있음)

`-` 만약 $B_k$를 처음이 아닌 나중에 주문한다면 $X_Q$가 $X^{s}_Q$보다 작으므로 $X_{Q+1} = X_Q + B_k$이다

`-` 만약 $B_k$를 처음에 주문한다면 어차피 $j,s$ 모두 처음이 아니게 되므로 새롭게 계산된 금액은 동일하다

`-` 따라서 $Q$개를 주문했을 때 최소 비용으로 주문할 수 있는 순서를 바탕으로 $X_{Q+1}$을 계산하지 않을 이유가 없다

`-` 이를 정리해보자

`-` 여태까지 주문한 음식들의 목록을 가지고 있자

`-` 첫 번째 음식을 변경하지 않을 때를 위해 $B_i$를 기준으로 오름차순 정렬한 배열 $R$을 가지고 있자

`-` 첫 번째 음식을 변경할 때를 위해 $A_i$를 기준으로 오름차순 정렬한 배열 $T$를 가지고 있자

`-` 처음에 $A_i$가 가장 작은 음식을 주문하며 시작했다

`-` 따라서 $i$번 음식을 주문한 음식들의 목록에 추가한 뒤 첫 번째 주문한 음식이 무엇인지 기억하고 있자

`-` 이후 $R$에서 주문 가능한 음식 중 첫 번째 음식을 주문하는 것과 $T$에서 주문하는 것을 비교할 것이다

`-` 이때 포인터를 사용하여 $R$과 $T$ 각각에서 주문 가능한 음식 중 가장 저렴한 것들을 가리키게 하자 

`-` $B_r$과 $A_t - A_{\operatorname{first}} + B_{\operatorname{first}}$ 중 더 작은 것을 선택하면 된다

`-` 주문한 음식의 번호를 주문 목록에 추가해야 한다

`-` 그리고 주문한 쪽 배열의 포인터를 오른쪽으로 한 칸 옮기자

`-` 이를 모든 음식을 주문할 때까지 반복하면 정답이다

`-` 참고로 $R$과 $T$는 음식들의 번호 순서가 다르게 배치되어 있다

`-` 따라서 현재 포인터가 가리키고 있는 음식이 이미 주문되어 있을 수 있다

`-` 그런 상황이라면 주문 가능한 음식이 나올 때까지 포인터를 한 칸 씩 오른쪽으로 옮겨주자

`-` 한 번 오른쪽으로 옮긴 포인터는 다시 왼쪽으로 이동하지 않으므로 최대 $N$칸만 움직일 수 있다

`-` 배열 $R$과 $T$를 만들기 위해 정렬하는 것은 $O(N \log N)$의 시간 복잡도를 가진다

`-` 한편, 이미 주문한 음식 목록에 대해 그 안에서 첫 번째 음식의 교체는 발생할 수 없다

`-` 즉, $1,3$번 음식을 이미 주문한 상태에서 나중에 음식을 추가하다보니 원래는 $1$번 음식을 처음 주문했다가 $3$번 음식을 처음으로 주문할 수 없다는 것이다

`-` 만약 그렇다면 애초에 $3$번 음식을 주문할 당시 $3$번 음식을 첫 번째로 주문했어야 한다

`-` 아무튼 전체 알고리즘의 시간 복잡도는 $O(N \log N)$이다

`-` 틀려서 생각해보니 배열 $T$를 $A_i - B_i$가 아닌 $A_i$를 기준으로 정렬해야 된다

`-` $A_i - B_i$는 $k$번 음식을 확정적으로 추가할 때 처음에 주문하거나 그렇지 않는 것을 기준으로 삼은 것이다

`-` 근데 두 경우는 중복될 수 없어서 무의미하다

`-` 또한 실제로 변화하는 금액은 $B_i$와 $A_k - A_j + B_j$ 중 하나이고 두 경우는 서로 다른 음식일 수 있다

`-` 따라서 $A_i$를 기준으로 정렬해야 된다 (위에서 말했지만 $- A_j + B_j$는 고정된 상수라 비교에 영향을 주지 않는다)

`-` $2$시간 $20$분 걸렸다

````python
def argmin(array):
    min_ = array[0]
    index = 0
    for i, a in enumerate(array):
        if a < min_:
            min_ = a
            index = i
    return index


def order_food_sequentially(first_prices, not_first_prices, R, T):
    first = argmin(first_prices)
    ordered = set([first])
    cost = first_prices[first]
    i_r = 0
    i_t = 0
    print(cost)
    while len(ordered) < N:
        while i_r < N and R[i_r] in ordered:
            i_r += 1
        while i_t < N and T[i_t] in ordered:
            i_t += 1
        B_r = not_first_prices[R[i_r]]
        A_t = first_prices[T[i_t]]
        if B_r <= A_t - first_prices[first] + not_first_prices[first]:
            cost += B_r
            ordered.add(R[i_r])
            i_r += 1
        else:
            cost += A_t - first_prices[first] + not_first_prices[first]
            ordered.add(T[i_t])
            first = T[i_t]
            i_t += 1
        print(cost)


def solution():
    global N, INF
    N = int(input())
    INF = float("inf")
    first_prices = []
    not_first_prices = []
    for i in range(N):
        A, B = map(int, input().split())
        first_prices.append(A)
        not_first_prices.append(B)
    R = list(range(N))  # B_i
    T = list(range(N))  # A_i
    R.sort(key=lambda x: not_first_prices[x])
    T.sort(key=lambda x: first_prices[x])
    order_food_sequentially(first_prices, not_first_prices, R, T)


solution()

# input
# 3
# 10 5
# 9 3
# 10 5
````

`-` 알고리즘 분류를 그리디가 아닌 동적 계획법으로 했다

`-` 둘 다 맞다고 생각하는데 나는 이전 항의 정보를 재사용한다는 것에 초점을 둬서 동적 계획법으로 태깅했다

## 기계오리 연구

- 문제 출처: [백준 28082번](https://www.acmicpc.net/problem/28082)

### 정답 풀이 (배낭 문제)

`-` 이제 골드 $1$ 문제를 풀어도 레이팅이 오르지 않지만 능지 이슈로 딱히 풀 문제가 없어서 풀어보려고 한다

`-` $N$개의 배터리 중 $K$개를 고르는 경우의 수는 $\binom{N}{K}$이고 $N$은 최대 $500$, $K$는 최대 $100$이므로 모든 경우의 수를 따지는 건 시간 초과이다

`-` 각 배터리의 전력량 $I_i$의 범위는 $1\le I_i \le 500$이다

`-` 전력량이 정수라는 조건이 없는데 정수라고 가정하겠다

`-` 정답을 맞힐려면 기계오리가 작동한 전력량의 종류의 수를 출력해야 하고 각 전력량도 오름차순으로 출력해야 한다

`-` 전력량이 정수가 아니면 대충 $\binom{500}{100}$개 이상의 전력량을 출력해야 되는데 말이 안된다

`-` 전력량이 정수라면 최대 $K$개의 배터리를 장착했을 때 가능한 전력량은 최소 $1$이며 최대 $500K$이다

`-` 임의의 전력량에 대해 해당 전력량을 $K$개 이하의 배터리를 조합하여 도출할 수 있는지 없는지 알면 될 것 같다

`-` $i$개의 배터리 중 $q$개 골라서 전력량 합으로 $x$를 만들고 또 $r$개를 골라서 전력량 합으로 $x$를 만들었다고 해보자

`-` $q <r$이라고 해보자

`-` 그러면 $r$개를 골라서 전력량 합으로 $x$를 만든 행위는 의미가 없다

`-` 왜냐하면 동일한 $i$개의 배터리 중 더 적은 $q$개만 선택해서 이미 전력량 합으로 $x$를 만들었기 때문이다

`-` 앞으로 고려할 $N-i$개의 배터리 중 선택할 수 있는 개수가 더 많아야 더 다양한 조합을 도출할 수 있다

`-` 여태까지 고려한 배터리로 만들 수 있는 전력량 합 $x$와 이를 만들기 위해 사용한 배터리의 개수를 기록하자

`-` 새로운 $I_i$의 전력량을 가진 배터리를 고려한다고 해보자

`-` 여태까지 만들어낸 전력량 합 $x$와 그 때의 배터리 개수 $j_x$가 기록되어 있을 것이다

`-` 그리고 이를 순회하며 $x$에 $I_i$를 더해주어 $j_{x+1}$개의 배터리를 사용해 새로운 전력량 합인 $x+I_i$를 만들 수 있다

`-` 만약 $x+ I_i$가 이미 고려한 배터리들로 만들어낸 전력량이며 이 때의 개수 $j_{x+I_i}$가 $j_{x+1}$보다 작다면 새로운 방법은 도움이 안 된다

`-` 처음엔 아무 배터리도 없으므로 $0$개의 배터리를 사용해 $0$의 전력량을 만든 정보만 가지고 있자

`-` 모든 배터리를 순회하는 건 $O(N)$이고 각 배터리마다 확인해야 하는 전력량의 종류의 개수는 최악의 경우 $O(KI)$이다

`-` 기계오리가 작동한 전력량을 형식에 맞춰 출력하기 위해 오름차순 정렬하는 건 $O(KI \log (KI))$이다

`-` 따라서 전체 알고리즘의 시간 복잡도는 $O(NKI + KI \log (KI))$이다

`-` $1$트 컷, $1$시간 $20$분 걸렸다

`-` 풀고 보니까 [양팔저울](https://www.acmicpc.net/problem/2629) 문제에 개수 제한이 추가된 버전의 문제였다

````python
def solution():
    N, K = map(int, input().split())
    I_list = list(map(int, input().split()))
    seen = {0: 0}  # 전력량: 사용한 배터리 개수
    for I in I_list:
        temp = {}
        for x, count in seen.items():
            if count >= K:
                continue
            temp[x + I] = min(count + 1, seen.get(x + I, K))
        seen.update(temp)
    print(len(seen) - 1)  # 0 제거
    print(*sorted(list(seen.keys()))[1:])


solution()

# input
# 3 2
# 1 100 10
````

`-` 다른 사람 풀이 보니까 딕셔너리 겁나 느리고 배열 써야 빠르다

## 우선순위와 쿼리

- 문제 출처: [백준 20892번](https://www.acmicpc.net/problem/28092)

### 정답 풀이 (분리 집합, 우선순위 큐)

`-` 제목에 우선순위가 들어가 있다

`-` 나는 우선순위하면 우선순위 큐가 떠오른다

`-` 이를 이용해서 풀 수 있을 것 같다

`-` 간선을 추가하는 연산은 분리 집합를 통해 구현하자

`-` 이때 $2$번 쿼리에 의해 정점의 개수가 가장 많은 트리가 선택되고 만약 개수가 같다면 더 작은 번호의 정점을 가진 트리가 선택된다

`-` 따라서 분리 집합을 구현할 때 루트 노드는 집합에 속한 정점 개수와 가장 작은 번호도 알아야 한다

`-` 참고로 그래프에 사이클이 있으면 트리가 아니다

`-` 따라서 union 연산을 수행할 때 두 노드가 이미 같은 그룹에 속해 있으면 두 그룹에 속한 모든 정점을 삭제해야 한다

`-` 굳이 모든 정점을 추적하는 대신 find 연산을 통해 루트 노드의 삭제 여부를 토대로 자신도 제거됐는지 판단하자

`-` 루트 노드가 제거됐다면 이와 연결된 자신도 제거된 것이다

`-` 삭제된 정점은 마킹을 해두어 삭제 여부를 $O(1)$에 알 수 있도록 하자

`-` $1$번 쿼리를 수행할 때 두 정점 중 하나라도 삭제된 정점이라면 나머지 정점 역시 삭제되야 한다

`-` 따라서 나머지 정점의 루트 노드를 추적해 삭제하도록 하자

`-` 한편, 우선순위가 높은 트리를 선택하기 위해 힙을 사용하자

`-` (크기, 가장 작은 번호, 정점 번호)를 튜플로 하여 힙에 넣을 것이다

`-` 이때 크기는 커야 좋으므로 앞에 마이너스 부호를 붙여서 사용하자 (파이썬은 기본적으로 최소 힙만 지원한다)

`-` 파이썬은 튜플을 비교할 때 앞에서부터 각 원소끼리의 크기를 비교한다 (이 문제 풀면서 처음 알았다, 맨 앞 원소만 비교하는 줄;)

`-` 트리의 크기는 겹칠지라도 번호는 유일하므로 우선순위가 동일한 경우는 없다

`-` 이러면 힙에 같은 정점이 여러 번 들어가긴 하는데 한 번 꺼내지면 삭제됐다고 마킹되니 괜찮다 (다익스트라 알고리즘 구현부와 비슷함)

`-` $2$번 쿼리가 입력으로 들어오면 가장 높은 우선순위의 원소를 꺼내자

`-` 만약 이미 제거된 정점이라면 제거되지 않은 정점이 나올 때까지 pop 연산을 수행하자 (트리가 없는데 $2$번 쿼리가 주어지는 경우는 없으니 힙에 최소 $1$개의 원소는 존재함)

`-` 원소를 꺼냈으면 트리에서 가장 번호가 작은 정점의 번호를 출력한 뒤 해당 정점의 루트 노드에 삭제했다는 마킹을 해두자

`-` $1$번 쿼리가 입력으로 들어오면 union 연산을 한 뒤 (크기, 가장 작은 번호, 루트 노드 번호)를 힙에 넣자

`-` 알고리즘의 시간 복잡도는 최악의 경우 $O(N + Q\log Q)$이다

`-` $1$시간 $40$분 걸렸다

`-` 분리 집합을 오랜만에 구현해서 $1$번 틀렸다 (개떡같이 구현했음)

````python
import heapq


def find(u):
    if parents[u] != u:
        parents[u] = find(parents[u])
    return parents[u]


def union(u, v):
    u_root = find(u)
    v_root = find(v)
    if u_root == v_root:
        removed[u_root] = True
        return
    if ranks[u_root] < ranks[v_root]:
        parents[u_root] = parents[v_root]
        sizes[v_root] += sizes[u_root]
        smallest_nums[v_root] = min(smallest_nums[u_root], smallest_nums[v_root])
    else:
        parents[v_root] = parents[u_root]
        sizes[u_root] += sizes[v_root]
        smallest_nums[u_root] = min(smallest_nums[u_root], smallest_nums[v_root])
        if ranks[u_root] == ranks[v_root]:
            ranks[u_root] += 1


def query1(heap, u, v):
    u_root = find(u)
    v_root = find(v)
    union(u, v)
    if removed[u_root] or removed[v_root]:
        removed[u_root] = True
        removed[v_root] = True
        return
    uv_root = find(u_root)
    heapq.heappush(heap, (-sizes[uv_root], smallest_nums[uv_root], uv_root))


def query2(heap):
    while heap:
        size, num, node = heapq.heappop(heap)
        root = find(node)
        if not removed[root]:
            print(smallest_nums[root])
            removed[root] = True
            break


def solution():
    global parents, ranks, removed, smallest_nums, sizes
    N, Q = map(int, input().split())
    parents = [u for u in range(N + 1)]
    ranks = [0] * (N + 1)
    smallest_nums = [u for u in range(N + 1)]
    sizes = [1] * (N + 1)
    removed = [False] * (N + 1)
    heap = [(-sizes[u], smallest_nums[u], parents[u]) for u in range(1, N + 1)]
    for _ in range(Q):
        info = list(map(int, input().split()))
        operator = info[0]
        if operator == 1:
            query1(heap, *info[1:])
        else:
            query2(heap)


solution()

# input
# 4 4
# 1 1 2
# 1 2 3
# 1 3 1
# 2
````

## 홀수 부분열

- 문제 출처: [백준 17279번](https://www.acmicpc.net/problem/17279)

### 정답 풀이 (배낭 문제)

`-` 랜덤 문제 고르는데 수학만 $4$번 연속 등장해서 어지러운 상태이다

`-` 배열의 크기 $n$이 최대 $60$이므로 가능한 모든 부분열을 고려하는 건 시간 초과이다

`-` 배낭 문제로 취급하여 해결할 수 있을 것 같다

`-` 배열 $A$의 각 원소는 $0$ 이상 $2000$ 이하의 정수이다

`-` 즉, 부분열의 합은 배열의 최댓값을 $s$라 할 때 $0$ 이상 $sn$ 이하의 정수이다

`-` 부분열의 원소들을 정렬했을 시 결과가 같다면 같은 부분열로 취급하니 중복해서 세지 않도록 주의하자

`-` 이를 위해 배열 $A$를 (원소, 개수) 형식의 튜플을 원소로 가지도록 변경할 것이다

`-` 즉, 배열 $A$에 있는 유일한 원소 $x$는 $(x,1)$로 변경되고 어떤 원소 $y$가 $k$개 있다면 이들을 하나로 합쳐 $(y, k)$로 치환된다

`-` $\operatorname{dp}[w]$를 합이 $w$인 부분열의 개수라고 하자

`-` 초깃값으로 $\operatorname{dp}[0] = 1$이다 (길이가 $1$인 부분열을 표현하기 위함)

`-` $0$은 어차피 홀수 부분열의 합이 아니라 나중에 정답에 카운팅 되지 않는다

`-` 현재 단계의 원소 $(y,k)$를 고려하자

`-` $0, y, \dots, ky$만큼 부분열의 합을 증가시킬 수 있다

`-` $w$는 $0$ 이상 $sn$ 이하로 큰 것부터 역순으로 $\operatorname{dp}$를 순회하며 증가시킬 부분열의 합을 $v$라 할 때 $\operatorname{temp}[w]$를 $\operatorname{dp}[w-v] + \operatorname{temp}[w]$로 갱신하자

`-` 단, $\operatorname{dp}[w-v] > 0$이어야 하는데 애초에 $0$이면 영향을 안 주니 신경쓰지 않아도 된다

`-` 이때 매 원소 $(y,k)$마다 현재 $\operatorname{dp}$ 배열을 임시 배열로 복사하고 임시 배열에 누적한 값을 순회가 끝난 뒤 $\operatorname{dp}$ 배열로 변경하자

`-` 이렇게 해야 부분열이 중복되지 않게 할 수 있다

`-` 배열 $A$의 마지막 원소까지 처리를 마쳤으면 $w$를 순회하며 $w$가 홀수 부분열의 합이라면 $\operatorname{dp}[w]$를 정답에 카운팅 해주자

`-` $\operatorname{dp}$ 배열을 채우는 건 최악의 경우 $O\left(sn^2\right)$이다

`-` $1$부터 $sn$까지 순회하며 자릿수 중 홀수인 것들의 개수가 홀수인지 판단하는 건 $O\left(sn\log_{10} sn\right)$이다

`-` 이를 테스트 케이스의 개수인 $t$번 반복하므로 전체 알고리즘의 시간 복잡도는 최악의 경우 $O\left(tsn\left(n + \log sn\right)\right)$이다

`-` $1$시간 $15$분 걸렸고 $1$트에 맞혔다

````python
from collections import defaultdict


def compress(array):
    dict_ = defaultdict(int)
    for a in array:
        dict_[a] += 1
    return list(dict_.items())


def knapsack(array, n):
    dp = [0] * (MAX_SUM + 1)
    dp[0] = 1
    for x, k in array:
        temp = dp.copy()
        for i in range(1, k + 1):
            v = x * i
            for w in range(MAX_SUM, -1, -1):
                if w - v < 0:
                    continue
                temp[w] += dp[w - v]
        dp = temp
    return dp


def is_odd_subset(x):
    c = 0
    while x > 0:
        if (x % 10) % 2 == 1:
            c += 1
        x //= 10
    return c % 2 == 1


def solve_testcase():
    global MAX_SUM
    n = int(input())
    A = list(map(int, input().split()))
    MAX_SUM = max(A) * n
    array = compress(A)
    dp = knapsack(array, n)
    count = sum(dp[x] for x in range(1, MAX_SUM + 1) if is_odd_subset(x))
    print(count)


def solution():
    t = int(input())
    for _ in range(t):
        solve_testcase()


solution()

# input
# 3
# 3
# 3 3 6
# 4
# 0 1 2 3
# 5
# 3 3 6 8 6
````
