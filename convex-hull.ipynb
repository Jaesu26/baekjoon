# 볼록 껍질 (Convex Hull)

`-` 모든 점을 포함하는 가장 작은 볼록 다각형을 만들자

## 볼록 껍질

- 문제 출처: [백준 1708번](https://www.acmicpc.net/problem/1708)

`-` 참고: https://en.wikipedia.org/wiki/Graham_scan

`-` $y$가 가장 작으면서 가장 왼쪽에 있는 점은 볼록 다각형을 구성하는 점에 항상 포함된다

`-` 해당 점을 기준으로 다른 점들을 반시계 방향으로 정렬하자

`-` 처음 기준 점부터 시작해 반시계 방향으로 정렬된 점들을 연결해 나가면 다각형이 만들어진다

`-` 이때 오목한 부분의 점을 제거하면 볼록 다각형이 된다

`-` 다각형에서 세 점 $A, B, C$가 연결됐다고 해보자

`-` 이때 점 $C$가 선분 $AB$의 오른쪽에 있다면 이들을 연결한 다각형은 오목해진다

`-` 점 $C$가 선분 $AB$의 왼쪽에 있다면 내각은 $180$도 이하이므로 볼록 다각형의 조건을 만족한다

`-` 여기서 점 $C$가 선분 $AB$의 왼쪽에 있는지 오른쪽에 있는지는 CCW를 이용해서 판단할 것이다

`-` 선분 $AB$와 선분 $BC$가 이루는 방향이 반시계 방향이라면 점 $C$는 선분 $AB$의 왼쪽에 있는 것이다

`-` 또한 점들을 반시계 방향으로 정렬하는 것도 CCW를 이용할 것이다

`-` CCW는 [CCW](https://www.acmicpc.net/problem/11758) 문제를 풀어서 알고 있다고 가정하겠다

`-` 가장 마지막으로 고려한 선분과 새로 추가할 점에 대해 해당 점이 선분의 오른쪽에 있다면 해당 선분은 볼록 다각형을 구성할 수 없다

`-` 그러므로 해당 선분을 제거한 뒤 점이 새로운 선분의 왼쪽에 있는지 확인하자

`-` 오른쪽에 있다면 이를 반복하면 된다

`-` 마지막 점까지 추가했다면 모든 점을 포함하는 볼록 다각형이 완성되며 이는 $O(N)$의 시간 복잡도를 가진다

`-` 정렬에 $O(N \log N)$을 소요하므로 컨벡스 헐을 계산하는 그레이엄 스캔의 시간 복잡도는 $O(N \log N)$이다

`-` 기벡 싫어~

```python
from functools import cmp_to_key


def ccw(p1, p2, p3):
    x1, y1 = p2[0] - p1[0], p2[1] - p1[1]
    x2, y2 = p3[0] - p2[0], p3[1] - p2[1]
    return x1 * y2 - x2 * y1


def dist(p1, p2):
    return (p1[0] - p2[0])**2 + (p1[1] - p2[1])**2


def compare(p1, p2):
    cross_product = ccw(p_base, p1, p2)
    if cross_product > 0:
        return -1
    if cross_product < 0:
        return 1
    if dist(p_base, p1) < dist(p_base, p2):
        return -1
    return 1


def solution():
    global p_base
    N = int(input())
    points = [list(map(int, input().split())) for _ in range(N)]
    points.sort(key=lambda x: (x[1], x[0]))
    p_base = points[0]
    points.sort(key=cmp_to_key(compare))
    stack = []
    for p in points:
        while len(stack) >= 2 and ccw(stack[-2], stack[-1], p) <= 0:
            stack.pop()
        stack.append(p)
    answer = len(stack)
    print(answer)


solution()

# input
# 8
# 1 1
# 1 2
# 1 3
# 2 1
# 2 2
# 2 3
# 3 1
# 3 2
```

## 단순 다각형

- 문제 출처: [백준 3679번](https://www.acmicpc.net/problem/3679)

`-` 컨벡스 헐을 만들 때 $y$가 가장 작고 가장 왼쪽에 있는 점을 기준으로 다른 점들을 반시계 방향으로 정렬한다

`-` 이들을 순서대로 연결하면 다각형을 얻을 수 있다

`-` 모든 점들을 이어야 하니 한 직선 위에 있는 점들에 대해 기준 점과 가까운 점부터 방문하자

`-` 이렇게 하면 선분이 서로 교차하지 않아 단순 다각형이 만들어진다

`-` 간단하게 확인하면 다음과 같다

`-` 연속된 두 점을 이은 선분 $e_i$은 기준 점에서 본 각도 구간 $[\theta_i, \theta_{i+1}]$에 속한다

`-` 각도 정렬을 수행했으므로 서로 다른 두 선분 $e_i, e_j$는 서로 교차하지 않는다

`-` 여기까지는 [볼록 껍질](https://www.acmicpc.net/problem/1708) 문제를 풀어서 알고 있던 내용이었다

`-` 그런데 예외가 하나 존재했다 ([볼록 껍질](https://www.acmicpc.net/problem/1708) 문제의 테케가 반례이다)

`-` 기준 점 $p_b$와 마지막으로 연결되는 점 $p_1$에 대해 어떤 점 $p_2$가 있을 때 $\operatorname{CCW}\left(\overrightarrow{p_b p_1}, \overrightarrow{p_b p_2}\right) = 0$이 성립될 수 있다

`-` 한 직선 위에 있는 점들에 대해 기준 점과 가까운 점부터 연결하기로 했으므로 $p_2$를 잇고 $p_1$을 잇는다

`-` 그런데 $p_2$를 잇고 $p_1$을 이은 다음 $p_b$와 연결하면 다각형이 완성되지 않는다 

`-` 그러므로 예외에 한해서 역으로 기준 점과 가장 먼 점을 먼저 방문해줘야 한다

`-` 이것 때문에 계속 틀려서 힘들었다

```python
from functools import cmp_to_key, partial


def dist(p1, p2):
    return (p1[0] - p2[0])**2 + (p1[1] - p2[1])**2


def ccw(p_base, p1, p2):
    return (p1[0] - p_base[0]) * (p2[1] - p_base[1]) - (p1[1] - p_base[1]) * (p2[0] - p_base[0])


def _compare(p_base, p1, p2):
    cross_product = ccw(p_base, p1, p2)
    if cross_product < 0:
        return 1
    if cross_product > 0:
        return -1
    if dist(p_base, p1) < dist(p_base, p2):
        return -1
    return 1


def handle_edge_case(points):
    n = len(points)
    index = -1
    for i in range(n - 2, -1, -1):
        if ccw(points[0], points[-1], points[i]) != 0:
            break
        index = i
    points[index:n] = points[index:n][::-1]
    return points


def solve_testcase():
    points = list(map(int, input().split()))
    n = points[0]
    points = [[points[2 * i + 1], points[2 * (i + 1)], i] for i in range(n)]
    points.sort(key=lambda p: (p[1], p[0]))
    compare = partial(_compare, points[0])
    points.sort(key=cmp_to_key(compare))
    points = handle_edge_case(points)
    order = [i for _, _, i in points]
    print(*order)


def solution():
    c = int(input())
    for _ in range(c):
        solve_testcase()


solution()

# input
# 2
# 4 0 0 2 0 0 1 1 0
# 5 0 0 10 0 10 5 5 -1 0 5
```

## 맹독 방벽

- 문제 출처: [백준 7420번](https://www.acmicpc.net/problem/7420)

`-` 방벽이 모든 건물들로부터 $L$ 이상 떨어져 있어야 한다는 조건이 없다고 해보자

`-` 그럼 모든 건물을 이은 다각형의 둘레를 따라 방벽을 세워볼 수 있다

`-` 그런데 해당 다각형이 오목 다각형이라면 오목한 부분을 없애고 연결하는 것이 둘레의 길이가 더 줄어든다

`-` 즉, 모든 건물을 포함하는 컨벡스 헐을 만들고 컨벡스 헐의 둘레만큼 방벽을 세우는 것이 방벽의 최소 길이이다

`-` 하지만 방벽은 모든 건물들로부터 $L$ 이상 떨어져야 한다

`-` 컨벡스 헐의 일부분인 연속된 점 $p_1,p_2,p_3$를 고려하자

`-` 컨벡스 헐의 선분 $\overline{p_1 p_2}$와 선분 $\overline{p_2 p_3}$는 일직선이다 (선분이니까 당연하다)

`-` 따라서 밖으로 $L$만큼 떨어진 곳에 일직선으로 동일한 길이의 방벽을 세우면 된다

`-` 해당 선분을 $\overline{p_a p_b}$와 $\overline{p_c p_d}$라고 하겠다

`-` 선분 $\overline{p_a p_b}$는 선분 $\overline{p_1 p_2}$와 평행하며 $L$만큼 떨어져 있다 ($\overline{p_c p_d}$와 $\overline{p_2 p_3}$도 마찬가지)

`-` 문제는 방벽이 점 $p_2$와도 $L$만큼 떨어지게 하기 위해 호를 고려해야 한다는 것이다

`-` 문제의 예시를 보면 알겠지만 맹독 방벽의 모퉁이는 호이다

`-` 호의 길이는 $L\theta$이므로 선분 $\overline{p_2 p_b}$와 선분 $\overline{p_2 p_c}$가 이루는 각 $\theta$를 알아야 한다

`-` 컨벡스 헐을 이루는 모든 점에 대해 호의 길이를 고려해줘야 한다

`-` 즉, 점 $p_i$에 대해 $L\theta_1 + L\theta_2 + \dots L\theta_n = L(\theta_1 + \theta_2 + \dots + \theta_n)$을 계산해야 된다

`-` 점 $p_i$의 내각을 $x_i$라고 하자

`-` $\theta_i = \pi - x_i$이므로 $n$각형($n \ge 3$)의 내각의 합은 $(n-2)\pi$임에 기인해 $\sum\limits_{i=1}^{n}\theta_i= n\pi - (n-2)\pi = 2\pi$이다

`-` 따라서 $\sum\limits_{i=1}^{n}L\theta_i = 2\pi L$이다

`-` 즉, 컨벡스 헐의 둘레의 길이에 $2\pi L$를 더한 값이 맹독 방벽의 최소 길이이다

`-` 컨벡스 헐의 둘레의 길이는 선분을 이루는 연속된 점들을 순회하며 유클리드 거리를 계산해 더해주면 된다

`-` 건물의 수를 $N$이라 할 때 그레이엄 스캔으로 컨벡스 헐을 구하는 건 $O(N \log N)$이고 선분을 순회하며 맹독 방벽의 길이를 계산하는 건 $O(N)$이다

`-` 따라서 전체 알고리즘의 시간 복잡도는 $O(N \log N)$이다

```python
import math
from functools import cmp_to_key, partial


def ccw(p1, p2, p3):
    return (p2[0] - p1[0]) * (p3[1] - p2[1]) - (p2[1] - p1[1]) * (p3[0] - p2[0])


def dist_square(p1, p2):
    return (p2[0] - p1[0])**2 + (p2[1] - p1[1])**2


def _compare(pivot, p1, p2):
    cross_product = ccw(pivot, p1, p2)
    if cross_product > 0:
        return -1
    if cross_product < 0:
        return 1
    if dist_square(pivot, p1) < dist_square(pivot, p2):
        return 1
    return -1


def graham_scan(points):
    stack = []
    for p in points:
        while len(stack) >= 2 and ccw(stack[-2], stack[-1], p) <= 0:
            stack.pop()
        stack.append(p)
    return stack


def measure_barrier_perimeter(convex_hull, dist_away_from):
    perimeter = 0
    for i in range(len(convex_hull) - 1):
        p1, p2 = convex_hull[i], convex_hull[i + 1]
        perimeter += dist_square(p1, p2)**0.5
    perimeter += dist_square(convex_hull[0], convex_hull[-1])**0.5
    perimeter += 2 * math.pi * dist_away_from
    return perimeter


def solution():
    N, L = map(int, input().split())
    points = [list(map(int, input().split())) for _ in range(N)]
    points.sort(key=lambda p: (p[1], p[0]))
    pivot = points[0]
    compare = partial(_compare, pivot)
    points.sort(key=cmp_to_key(compare))
    convex_hull = graham_scan(points)
    answer = measure_barrier_perimeter(convex_hull, L)
    print(f"{answer:.0f}")


solution()

# input
# 9 100
# 200 400
# 300 400
# 300 300
# 400 300
# 400 400
# 500 400
# 500 200
# 350 200
# 200 200
```

## 로버트 후드

- 문제 출처: [백준 9240번](https://www.acmicpc.net/problem/9240)

`-` 가장 먼 두 점은 볼록 껍질을 구성하는 점이어야 한다

`-` 그렇지 않고 볼록 껍질 내부의 점이라면 볼록 껍질 위의 점을 골라 거리를 더 멀게 할 수 있다

`-` 어느 한 점이 볼록 껍질 내부의 점이라고 하자

`-` 해당 점까지 선분을 긋고 해당 점에서 그 선분과 수직인 직선을 고려하자 

`-` 그럼 그 직선 위 또는 직선을 지나야 만날 수 있는 볼록 껍질 위의 점이 존재하고 해당 점까지의 거리가 더 멀게 된다

`-` 따라서 볼록 껍질을 구성하는 점들 중 두 점을 골라 가장 먼 두 점을 고를 수 있다

`-` 모든 조합을 확인하는 방법은 볼록 껍질을 구성하는 점의 개수를 $K$라 할 때 $O\left(K^2\right)$의 시간 복잡도를 가진다

`-` 그럼 전체 알고리즘의 시간 복잡도는 $O\left(N\log N+K^2\right)$이다

`-` 그런데 좌표는 정수이고 절댓값은 $1000$을 넘지 않는다

`-` 단순하게 외곽의 모든 점을 볼록 껍질에 포함시켜도 $K$는 최대 $8000$이므로 제한 시간 안에 동작할 수 있다

```python
from functools import cmp_to_key, partial


def ccw(p1, p2, p3):
    return (p2[0] - p1[0]) * (p3[1] - p2[1]) - (p2[1] - p1[1]) * (p3[0] - p2[0])


def compute_distance_square(p1, p2):
    return (p2[0] - p1[0])**2 + (p2[1] - p1[1])**2


def _compare(p_base, p1, p2):
    cross_product = ccw(p_base, p1, p2)
    if cross_product > 0:
        return -1
    if cross_product < 0:
        return 1
    d1 = compute_distance_square(p_base, p1)
    d2 = compute_distance_square(p_base, p2)
    if d1 > d2:
        return 1
    return -1


def graham_scan(points):
    stack = []
    for p in points:
        while len(stack) >= 2 and ccw(stack[-2], stack[-1], p) <= 0:
            stack.pop()
        stack.append(p)
    return stack


def compute_convex_hull_diameter(convex_hull):
    diameter = 0
    for i in range(1, len(convex_hull)):
        for j in range(i):
            d = compute_distance_square(convex_hull[i], convex_hull[j])
            diameter = max(d, diameter)
    return diameter**0.5
 

def solution():
    C = int(input())
    points = [list(map(int, input().split())) for _ in range(C)]
    p_base = min(points, key=lambda p: (p[1], p[0]))
    compare = partial(_compare, p_base)
    points.sort(key=cmp_to_key(compare))
    convex_hull = graham_scan(points)
    answer = compute_convex_hull_diameter(convex_hull)
    print(answer)


solution()

# input
# 2
# 2 2
# -1 -2
```

## 가장 먼 두 점

- 문제 출처: [백준 2049번](https://www.acmicpc.net/problem/2049)

`-` [로버트 후드](https://www.acmicpc.net/problem/9240) 문제에서 좌표의 범위가 커진 버전이다

`-` 볼록 껍질을 구성하는 점 중 가능한 모든 조합을 확인하여 가장 먼 두 점을 알 수 있다

`-` $N$은 전체 점의 개수이고 $K$는 볼록 껍질을 구성하는 점의 개수이다

`-` 그럼 해당 알고리즘의 시간 복잡도는 $O\left(N\log N+K^2\right)$이다

`-` 그런데 좌표는 정수이고 절댓값은 $10000$을 넘지 않는다

`-` 단순하게 외곽의 모든 점을 볼록 껍질에 포함시키면 $K$는 최대 $80000$이다

`-` 그런데 연속한 점들이 같은 직선 위의 존재하면 양 끝 점만 취하므로 실제 $K$는 이보다 작다

`-` 이로 인해 위의 알고리즘이 제한 시간 안에 동작할 수 있다

```python
from functools import cmp_to_key, partial


def ccw(p1, p2, p3):
    return (p2[0] - p1[0]) * (p3[1] - p2[1]) - (p2[1] - p1[1]) * (p3[0] - p2[0])


def compute_distance_square(p1, p2):
    return (p2[0] - p1[0])**2 + (p2[1] - p1[1])**2


def _compare(p_base, p1, p2):
    cross_product = ccw(p_base, p1, p2)
    if cross_product > 0:
        return -1
    if cross_product < 0:
        return 1
    d1 = compute_distance_square(p_base, p1)
    d2 = compute_distance_square(p_base, p2)
    if d1 > d2:
        return 1
    return -1


def graham_scan(points):
    stack = []
    for p in points:
        while len(stack) >= 2 and ccw(stack[-2], stack[-1], p) <= 0:
            stack.pop()
        stack.append(p)
    return stack


def compute_convex_hull_diameter_square(convex_hull):
    diameter_square = 0
    for i in range(1, len(convex_hull)):
        for j in range(i):
            d = compute_distance_square(convex_hull[i], convex_hull[j])
            diameter_square = max(d, diameter_square)
    return diameter_square
 

def solution():
    C = int(input())
    points = [list(map(int, input().rstrip().split())) for _ in range(C)]
    p_base = min(points, key=lambda p: (p[1], p[0]))
    compare = partial(_compare, p_base)
    points.sort(key=cmp_to_key(compare))
    convex_hull = graham_scan(points)
    answer = compute_convex_hull_diameter_square(convex_hull)
    print(answer)


solution()

# input
# 4
# 1 0
# 1 1
# 0 0
# 0 1
```
