# 스프라그–그런디 정리 (Sprague–grundy Theorem)

`-` 일반적인 규칙하에 모든 공정한 게임은 단일 힙 님 게임 또는 님의 무한 일반화와 동치이다

`-` 따라서 게임판의 각 상태는 하나의 자연수(그런디 수)로 나타낼 수 있다

`-` 참고: https://en.wikipedia.org/wiki/Sprague-Grundy_theorem

## 님 게임 2

- 문제 출처: [백준 11868번](https://www.acmicpc.net/problem/11868)

`-` 참고: https://en.wikipedia.org/wiki/Nim#Proof_of_the_winning_formula

```python
def solution():
    N = int(input())
    P = list(input().split())
    grundy_num = eval("^".join(P))
    if grundy_num == 0:
        print("cubelover")
        return
    print("koosaga")


solution()

# input
# 4
# 1 2 3 4
```

## 님 게임

- 문제 출처: [백준 11694번](https://www.acmicpc.net/problem/11694)

`-` 모든 돌 더미가 $1$개의 돌을 가지고 있을 때 돌 더미가 개수가 짝수이면 선공이 이기고 홀수이면 후공이 이긴다

`-` 이를 제외하면 [님 게임 2](https://www.acmicpc.net/problem/11868)와 풀이법이 같다

```python
def solution():
    N = int(input())
    P = list(input().split())
    if all(p == "1" for p in P):
        if N % 2 == 0:
            print("koosaga")
            return
        print("cubelover")
        return
    grundy_num = eval("^".join(P))
    if grundy_num == 0:
        print("cubelover")
        return
    print("koosaga")


solution()

# input
# 4
# 1 2 3 4
```
