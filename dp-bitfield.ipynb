# 비트필드를 이용한 동적 계획법 (Dynamic Programming Using Bitfield)

## 계단 수

- 문제 출처: [백준 1562번](https://www.acmicpc.net/problem/1562)

`-` 삽질을 너무 많이 해서 질문 검색이랑 이전에 해결한 [쉬운 계단 수](https://www.acmicpc.net/problem/10844) 풀이 보고옴

`-` 문제 풀이에 앞서 삽질한 내용을 간략히 정리하겠다

`-` $0$부터 $9$까지 전부 사용해서 계단 수를 만들어야 하므로 $N=10$일 때 유일한 계단 수 $9876543210$이 존재한다

`-` $N=11$인 계단 수를 만들 때 앞 또는 뒤에 $1$차이 나는 숫자 하나를 추가하여 만든다 (이러면 $N=12$일 때 $109876543210$은 못 만듦 ㅋㅋ)

`-` 이 과정은 다이나믹 프로그래밍으로 해결 가능하다

`-` 하지만 큰 문제가 있다...

`-` 위의 방식대로 풀면 중복 카운팅되는 계단 수가 존재한다 (그리고 만들어지지 않는 계단 수도 존재한다...)

`-` 한 방향으로만 숫자를 추가해야 하는데 앞과 뒤 아무데나 추가할 수 있으므로 유일성이 보장되지 않는다

`-` 예를 들어 숫자를 뒤에만 추가한다고 해보자

`-` 맨 앞에 숫자가 $1$인 것과 그렇지 않은 것이 있을텐데 무슨 짓을 해도 그렇지 않은 숫자를 맨 앞에 숫자가 $1$인 계단 수로 만들 수 없다

`-` 왜냐하면 숫자를 뒤에만 추가하기 때문이다 (쉬운 계단 수 문제 풀이에 사용한 방법임)

`-` 하지만 숫자를 앞에 추가할 수 있으면 맨 앞에 숫자를 어느정도 추가하면 맨 앞이 $1$인 계단 수도 만들 수 있다

`-` 이제 삽질 내용은 다 말했으니 본격적으로 문제를 풀어보자

`-` 일단 아이디어는 쉬운 계단 수 문제를 풀 때와 동일하다

`-` 기존 계단 수 마지막에 숫자를 하나 추가해서 새로운 계단 수를 만드는 방식으로 할 것이다

`-` 문제는 $0$부터 $9$까지 전부 사용해야 한다는 것이다

`-` 다이나믹 프로그래밍을 통해 계단 수를 카운팅하면서 $0$부터 $9$ 중 어느것을 사용했는지도 같이 기록해야 한다

`-` 임의의 계단 수는 $0$부터 $9$ 각각에 대해 포함하고 있거나 그렇지 않다

`-` 즉, 각 숫자에 포함여부를 $0$ 또는 $1$로 나타내고 이를 이어 붙인다면 각 숫자의 포함여부를 쉽게 판단할 수 있다

`-` 이렇게 만든 수는 $0$과 $1$만 사용하므로 이진법으로 나타낸 것이며 이를 십진법으로 변환할 수도 있다

`-` 십진법으로 변환하게 되면 최댓값은 $0$부터 $9$까지 모든 숫자를 사용한 $1111111111$이고 이를 십진법으로 변환시 $1023$이다

`-` 현재 계단 수의 길이와 마지막 숫자, 그리고 $0$부터 $9$까지 숫자의 포함여부를 나타낸 수를 바탕으로 점화식을 세울 수 있다

`-` 쉬운 계단 수와 동일한데 추가할 마지막 숫자를 비트마스크에 기록하면 된다

`-` 만약 마지막 숫자가 $2$이고 현재 계단 수에 $2$가 포함되어 있지 않다면 $2$에 해당하는 위치의 비트를 $1$로 바꿔야 한다

`-` $1$을 왼쪽으로 $2$칸 옮기면 $100$이 되고 이는 숫자 $4$이다

`-` 기존 계단 수와 숫자 $4$에 or 연산을 취하면 $2$에 해당하는 위치의 비트가 $1$이 된다

`-` 결과적으로 정답은 길이가 $N$이고 마지막 숫자가 $0$부터 $9$인 계단 수 중 비트 변환수가 $1023$인 것들의 합이다

```python
N = int(input())
P = int(1e9)
# dp[n][k][b]는 k로 끝나는 n자리 계단 수 중 비트 변환수가 b인 것들의 총 개수
dp = [[[0 for _ in range(1024)] for _ in range(10)] for _ in range(N + 1)]
for k in range(1, 10):
    dp[1][k][1 << k] = 1  # 1 ~ 9는 계단 수이다
for n in range(2, N + 1):
    for k in range(10):
        for b in range(1024):
            b_new = b | (1 << k)
            if k > 0:
                dp[n][k][b_new] += dp[n - 1][k - 1][b]
            if k < 9:
                dp[n][k][b_new] += dp[n - 1][k + 1][b]
            dp[n][k][b_new] %= P
answer = 0
for i in range(10):
    answer += (dp[N][i][1023] % P)
answer %= P
print(answer)

# input
# 10
```

`-` 질문검색 추가로 봤는데 비트마스킹을 사용하지 않은 풀이도 있다 (약간 콜럼버스 달걀 느낌임)

`-` $0$부터 $9$까지 모두 사용한 계단 수만 고려해야 한다

`-` 모두 사용해야 하므로 어떤 숫자가 사용됐는지 여부를 고려할 필요없이 사용된 숫자의 최소, 최대만 알아도 된다

`-` 계단 수이므로 최솟값과 최댓값은 연결될 수밖에 없다 (연결이 불가능하면 계단 수가 될 수 없음, 중간에 차이가 $1$을 넘어간다는 의미임)

`-` 이걸 이용하여 자릿수, 최솟값, 최댓값, 마지막 자리 숫자를 4차원 배열로 만들어 dp로 해결할 수 있다

`-` 결과적으로 `dp[N][0][9][0] + ... + dp[N][0][9][9]`가 정답이 된다

## 할 일 정하기1

- 문제 출처: [백준 1311번](https://www.acmicpc.net/problem/1311)

`-` 사람마다 하나의 일을 무조건 수행해야 한다

`-` 이 문제를 단순하게 생각하면 $N!$의 시도로 정답을 찾을 수 있지만 $N$이 최대 $20$이므로 시간 초과이다

`-` $f(X, S)$를 집합 $X$에 속한 사람들이 집합 $S$의 일을 수행할 때 드는 비용의 최솟값이라고 하자

`-` 어차피 사람마다 하나의 일을 수행해야 하므로 $1$번부터 순서대로 $N$번까지 고려하겠다

`-` 그럼 $f(\{1,2,\cdots,N\}, \{1,2,\cdots,N\}) = \min(\{f(\{1\}, \{s\}) + f(\{2,\cdots, N\}, \{1, \cdots, N\} - \{s\})\; |\; s=1,2,\cdots,N\})$이다

`-` 다시 $f(\{2,\cdots, N\}, \{1, \cdots, N\} - \{s_{\min}\})$ 대해서 재귀적으로 계산하면 된다

`-` base case로 한 명만 남았다면 맡을 일도 하나이므로 해당 일을 하는데 드는 비용을 반환하면 된다

`-` 담당할 일을 집합으로 관리하게 되면 집합을 만드는데 $O(N)$의 시간 복잡도와 공간 복잡도를 가지게 된다

`-` 대신 비트마스크로 저장하여 $O(1)$에 처리하자

`-` `dp[n][b]`를 $1$번부터 $n$번 사람까지 일을 부여했고 수행한 일의 비트마스크 이진수가 $b$일 때 비용의 최솟값이라고 하자

`-` 초기값으로 $n=1$인 경우 $n$개의 일 중 하나의 일을 선택하여 처리하는 비용은 $D_{1i}$이며 이것이 최솟값이다

`-` 그리고 비트마스크에서 해당 일의 번호에 해당하는 위치의 비트를 $0$에서 $1$으로 만들면 된다

`-` $n=2$부터는 전의 상태에서 가능한 할 수 있는 일을 수행한 뒤 해당 번호의 비트를 키면 되며 동일한 $n$과 $b$에 대해 최솟값을 갱신할 수 있다면 갱신한다

`-` 이를 $n=N$까지 반복하면 된다

`-` 최종적으로 `dp[N][2^N - 1]`이 모든 일을 하는데 필요한 비용의 최솟값이 된다

`-` `dp`가 가질 수 있는 상태는 $\binom{N}{1} + \binom{N}{2} + \cdots + \binom{N}{N}$이므로 총 $2^N$이다

`-` $i$번째 사람이 $N$개의 일 중 어떤 일을 수행할 수 있는지 확인하기 위해 `dp[i - 1]`을 순회하므로 알고리즘의 시간 복잡도는 $O\left(N2^N\right)$이다

`-` 만약 비트마스크가 아닌 집합으로 현재 남은 일을 관리했다면 반복문마다 집합을 복사하기 위해 $O(N)$의 비용이 추가로 드므로 시간 초과이다

```python
def assign_work(cost, n):
    b_max = 2**n - 1
    dp = [{} for _ in range(n + 1)]  # dp[n][b]는 1번부터 n번까지 일을 했고 한 일의 비트 변환수가 b일 때 비용의 최솟값
    for w in range(1, n + 1):
        b = 1 << (w - 1)  # 1 ~ 2^(N-1)
        dp[1][b] = cost[0][w - 1]
    for i in range(2, n + 1):
        for j in range(1, n + 1):  # i번 사람이 j번 일을 수행
            for b, w in dp[i - 1].items():
                b_new = b | (1 << (j - 1))
                if b == b_new:
                    continue
                dp[i][b_new] = min(cost[i - 1][j - 1] + w, dp[i].get(b_new, INF))
    return dp[n][b_max]


def solution():
    global INF
    N = int(input())
    cost = [list(map(int, input().split())) for _ in range(N)]
    INF = float("inf")
    answer = assign_work(cost, N)
    print(answer)


solution()

# input
# 3
# 2 3 3
# 3 2 3
# 3 3 2
```

## 외판원 순회

- 문제 출처 [백준 2098번](https://www.acmicpc.net/problem/2098)

`-` 단순하게 모든 경우의 수를 탐색하는 건 $N!$의 시도가 필요한데 이는 최대 $16!$이므로 시간 초과이다

`-` 비트마스크와 dfs를 결합하여 문제를 해결할 수 있다

`-` 여태까지 비용을 들여 방문한 도시를 비트마스크로 관리하자

`-` 이전의 방문한 도시는 제외하고 현재 도시와 인접한 도시를 방문하는 것을 고려하자

`-` 출발지는 무조건 마지막에 다시 방문해야 하므로 여태까지 방문한 도시뿐만 아니라 출발지도 고려해야 한다

`-` 그런데 어차피 모든 도시를 방문해야 하고 이는 원순열과 마찬가지이다

`-` 그러니 출발지를 임의로 고정해도 괜찮다

`-` 방문 루트가 사이클을 이루므로 회전시키면 출발지가 바뀐다

`-` 즉, 방문하는 모든 경우의 수를 만들 수 있다

`-` $1$번 도시를 출발지로 지정하자

`-` 여태까지 방문한 도시가 같더라도 현재 위치하고 있는 도시가 다르면 결과도 다를 수 있다 

`-` 현재 위치한 도시와 방문 도시 목록이 동일할 때 비용이 더 적으면 탐색을 계속하자

`-` 비용을 들여 방문한 도시의 수를 $n$이라 할 때 $n = N - 1$이면 다음 방문지는 출발지이며 이것이 마지막 방문이다 

`-` 현재 위치한 도시의 경우의 수는 $N$이며 비트마스크의 크기는 $2^N$이다

`-` 위와 같이 했다가 틀려서 원인을 파악했다

`-` 위의 알고리즘은 비용이 더 적을 때만 방문한다

`-` 원래의 방문 횟수는 $O(N!)$이지만 비용이 더 적을 때만 방문하므로 $O\left(N^2 2^N\right)$라고 생각했다

`-` 그런데 항상 비용을 갱신하는 최악의 경우엔 방문 횟수가 $O(N!)$이므로 시간 초과이다

`-` 대신 이전의 $n-1$ 상태로부터 $n$번째에 $x$ 도시를 방문한다고 생각해보자

`-` $n-1$개의 도시를 방문한 상태에서 $x$ 도시를 방문하지 않은 상태에 대해 $x$ 도시를 방문했을 때 누적 비용을 최소로 하는 것을 찾자

`-` 그러면 $O\left(N^2 2^N\right)$에 처리가 가능하므로 제한 시간 안에 통과할 수 있다

```python
def tsp(graph):
    # dp[n][x][b]는 n번째에 x에 위치하고 방문 도시 비트 변환수가 b일 때 방문 비용의 최솟값
    dp = [[{} for _ in range(N + 1)] for _ in range(N + 1)]
    # 초깃값
    for u in range(1, N + 1):
        w = graph[START][u]
        if w == 0:
            continue
        dp[1][u][1 << (u - 1)] = w
    for i in range(2, N + 1):
        for current in range(1, N + 1):
            if current == START and i < N:
                continue
            for prev in range(1, N + 1):
                w = graph[prev][current]
                if w == 0:
                    continue
                for bit_prev, cost in dp[i - 1][prev].items():
                    bit_now = bit_prev | (1 << (current - 1))
                    if bit_prev == bit_now:
                        continue
                    dp[i][current][bit_now] = min(cost + w, dp[i][current].get(bit_now, INF))
    return dp


def solution():
    global N, INF, START
    N = int(input())
    MAX_BIT = 2**N - 1
    graph = [[0 for _ in range(N + 1)]]
    for _ in range(N):
        weights = [0] + list(map(int, input().split()))
        graph.append(weights)
    INF = float("inf")
    START = 1
    dp = tsp(graph)
    answer = dp[N][START][MAX_BIT]
    print(answer)


solution()

# input
# 4
# 0 10 15 20
# 5 0 9 10
# 6 13 0 12
# 8 8 9 0
```

## 박성원

- 문제 출처: [백준 1086번](https://www.acmicpc.net/problem/1086)

`-` 단계별로 풀어보기에 있는 비트마스킹 동적 계획법 문제이다

`-` $N$개의 수를 가지고 합친수를 만드는 방법은 총 $N!$개로 $N$이 최대 $15$이니 완전 탐색은 시간 초과이다

`-` 잘 생각해보면 우리에게 필요한 건 합친수가 정수 $K$로 나누어 떨어지냐는 것이다

`-` 즉, 합친수를 굳이 알 필요 없이 $K$로 나눈 나머지만 알면 된다

`-` 어떤 합친수 $A$가 있고 $A$ 뒤에 $x$를 이어 붙여 새로운 합친수 $B$를 만든다고 해보자

`-` $x$의 자릿수를 $d$라 하면 $B = A \times 10^d + x$이다 ($d$는 최대 $50$이다)

`-` 따라서 $B \operatorname{mod} K = ((A \operatorname{mod} K) \times (10^d \operatorname{mod} K) + (x \operatorname{mod} K)) \operatorname{mod} K$이다

`-` $10^d \operatorname{mod} K$와 $x \operatorname{mod} K$는 미리 계산해 놓으면 $O(1)$에 처리 가능하므로 $A \operatorname{mod} K$만 알면 $B \operatorname{mod} K$를 $O(1)$에 계산할 수 있다

`-` $10^d \operatorname{mod} K$는 분할 정복을 통해 $O(\log d)$에 계산 가능하며 $x$의 자릿수는 문자열로 바꾸고 길이를 재서 $O\left(d^3\right)$에 계산 가능하다

`-` $n$개의 숫자를 가지고 만들 수 있는 모든 합친수에 대해 이를 $K$로 나눈 나머지와 등장 횟수를 딕셔너리로 관리하자

`-` $n$개의 수를 합쳐서 만들어진 수의 집합을 $A_n$이라 하고 임의의 원소를 $A$라 하자

`-` $\operatorname{dp}[n][b]$는 사용된 $n$개의 수 집합의 비트 변환수가 $b$일 때 가능한 모든 $A$에 대해 $A$를 $K$로 나눈 나머지의 등장 횟수를 기록한 집합이라 하자

`-` 그럼, $A_n$의 임의의 원소 $A$에 대해 $A \operatorname{mod} K$인 $\operatorname{dp}[n - 1][b_{\operatorname{prev}}]$를 바탕으로 $B_n$의 임의의 원소 $B$에 대해 $B \operatorname{mod} K$인 $\operatorname{dp}[n][b_{\operatorname{now}}]$를 계산할 수 있다

`-` 원래라면 $N!$의 경우의 수를 탐색해야 했지만 위와 같이 메모이제이션을 활용하면 훨씬 적은 시간에 탐색 가능하다

`-` 가능한 비트 변환수는 $2^N$개 존재하며 $\operatorname{dp}[n][b]$의 크기는 $K$이다

`-` 비트 변환수를 바탕으로 어떤 숫자가 사용되는지 탐색하는데 $O(N)$의 시간 복잡도를 가진다

`-` 따라서 $O(N!)$ 대신에 $O\left(KN2^N\right)$의 시간 복잡도로 해결 가능하다

`-` 최종적으로 $\operatorname{dp}[N][2^N - 1][0]$이 박성원이 정답을 맞힐 수 있는 랜덤 순열의 개수이다

`-` $\dfrac{\operatorname{dp}[N][2^N - 1][0]}{N!}$을 기약분수로 나타내면 정답이다

`-` 코랩에서 코드 수정한 것 때문에 AI가 `dp[n][bit_new][mod_new] += count` 부분을 자동완성 해버려서 스포당했다

`-` 근데 내가 로직 작성한 노트에 `+= count`를 작성해두어서 어차피 올바르게 수정할 운명이긴 했다

```python
from collections import defaultdict


def fact(n):
    if n <= 1:
        return 1
    return n * fact(n - 1)


def compute_gcd(a, b):
    r = a % b
    if r == 0:
        return b
    return compute_gcd(b, r)


def compute_digit(x):
    return len(str(x))


def power(a, b, c):
    a = a % c
    if b <= 1:
        return a
    half = power(a, b // 2, c)
    result = (half % c)**2 % c
    if b % 2 == 0:
        return result
    return (a * result) % c


def to_irreducible_fraction(numerator, denominator):
    gcd = compute_gcd(denominator, numerator)
    p = numerator // gcd
    q = denominator // gcd
    return p, q


def make_answer(numerator, denominator):
    if numerator == 0:
        answer = "0/1"
    elif numerator == denominator:
        answer = "1/1"
    else:
        p, q = to_irreducible_fraction(numerator, denominator)
        answer = f"{p}/{q}"
    return answer


def bottom_up(array, digits, power_mods):
    dp = [defaultdict(dict) for _ in range(N + 1)]
    for i, x in enumerate(array):
        dp[1][1 << i][x % K] = 1
    for n in range(2, N + 1):
        for i, x in enumerate(array):
            d = digits[i]
            p = power_mods[i]
            for bit_prev in dp[n - 1]:
                bit_now = bit_prev | (1 << i)
                if bit_now == bit_prev:
                    continue
                for mod_prev, count in dp[n - 1][bit_prev].items():
                    mod_now = (mod_prev * p + x % K) % K
                    if mod_now not in dp[n][bit_now]:
                        dp[n][bit_now][mod_now] = count
                    else:
                        dp[n][bit_now][mod_now] += count
    return dp


def solution():
    global N, K
    N = int(input())
    array = [int(input()) for _ in range(N)]
    K = int(input())
    digits = [compute_digit(array[i]) for i in range(N)]
    power_mods = [power(10, digits[i], K) for i in range(N)]
    MAX_BIT = 2**N - 1
    dp = bottom_up(array, digits, power_mods)
    numerator = dp[N][MAX_BIT].get(0, 0)
    denominator = fact(N)
    answer = make_answer(numerator, denominator)
    print(answer)


solution()

# input
# 3
# 3
# 2
# 1
# 2
```

## 007

- 문제 출처: [백준 3056번](https://www.acmicpc.net/problem/3056)

`-` [할 일 정하기 1](https://www.acmicpc.net/problem/1311) 문제를 풀었으면 이 문제도 해결할 수 있다

```python
def solution():
    N = int(input())
    man2probs = [list(map(lambda x: int(x) / 100, input().split())) for _ in range(N)]
    MAX_BIT = 2**N - 1
    dp = [0] * (MAX_BIT + 1)  # dp[b]는 수행한 임무 n개의 상태가 b일 때 지미 본드 1부터 지미 본드 n까지 해당 임무를 수행해서 성공할 최대 확률 (n은 비트에서의 1의 개수)
    for mission in range(N):
        dp[1 << mission] = man2probs[0][mission]  
    for b in range(1, MAX_BIT):
        n = bin(b).count("1")
        for mission in range(N):
            b_next = b | (1 << mission)
            if b == b_next:
                continue
            dp[b_next] = max(dp[b] * man2probs[n][mission], dp[b_next])
    print(dp[MAX_BIT] * 100)


solution()

# input
# 3
# 25 60 100
# 13 0 50
# 12 70 90
```

## 발전소

- 문제 출처: [백준 1102번](https://www.acmicpc.net/problem/1102)

`-` 초깃값으로 켜진 발전소를 고려하자

`-` 결국엔 모든 발전소를 켜야하는데 어떤 발전소를 켜야되는지 모른다

`-` 따라서 다 해봐야 한다

`-` 꺼진 발전소 중 하나를 고른다

`-` 켜진 발전소 중 하나를 골라 가장 적은 비용으로 꺼진 발전소를 켤 수 있도록 한다

`-` 모든 발전소가 켜질 때까지 반복하자

`-` 초기에 모든 발전소가 꺼진 상태가 아니라면 모든 발전소를 켤 수 있다

`-` 발전소의 가동 여부를 비트마스크로 관리하자

`-` 모든 상태 공간의 크기는 $O\left(2^N\right)$이다

`-` 각 상태 공간에 대해 어떤 발전소를 켤지 정하자

`-` 이는 $O(N)$의 시간 복잡도를 가진다

`-` 어떤 발전소를 켤지 정했으면 비용을 정하기 위해 켜진 발전소 중 하나를 골라야 하는데 이는 $O(N)$의 시간 복잡도를 가진다

`-` 따라서 전체 알고리즘의 시간 복잡도는 $O\left(N^2 2^N\right)$이다

```python
def select_cost(on_powers, off_power, power2costs):
    cost = INF
    for p in on_powers:
        if power2costs[p][off_power] < cost:
            cost = power2costs[p][off_power]
    return cost


def get_on_powers(bit):
    on_powers = []
    for p in range(N):
        if bit == bit | (1 << p):
            on_powers.append(p)
    return on_powers


def bitmask_dp(power2costs, power2is_on):
    dp = [INF] * (MAX_BIT + 1)
    init_bit = int("".join(reversed(power2is_on)), 2)
    if bin(init_bit).count("1") >= P:
        return 0
    if init_bit == 0:
        return -1
    dp[init_bit] = 0
    answer = INF
    for b in range(MAX_BIT):
        if dp[b] == INF:
            continue
        for power_next in range(N):
            b_next = b | (1 << power_next)
            if b == b_next:
                continue
            on_powers = get_on_powers(b)
            cost = select_cost(on_powers, power_next, power2costs)
            dp[b_next] = min(dp[b] + cost, dp[b_next])
            n = bin(b_next).count("1")
            if n >= P and dp[b_next] < answer:
                answer = dp[b_next]
    return answer


def solution():
    global N, P, INF, MAX_BIT
    N = int(input())
    power2costs = [list(map(int, input().split())) for _ in range(N)]
    power2is_on = ["1" if is_on == "Y" else "0" for is_on in list(input())]
    P = int(input())
    INF = float("inf")
    MAX_BIT = 2**N - 1
    answer = bitmask_dp(power2costs, power2is_on)
    print(answer)


solution()

# input
# 3
# 0 2 4
# 2 0 3
# 4 3 0
# YNN
# 3
```

## 끝말잇기

- 문제 출처: [백준 2320번](https://www.acmicpc.net/problem/2320)

`-` 비트필드를 활용한 동적계획법이다

`-` 똑같은 단어를 사용했어도 끝말이 뭐인지가 중요하므로 끝말을 고려해야 한다

`-` 사용한 단어가 아닌 단어 중 끝말잇기가 가능하다면 끝에 붙이자

`-` 사전의 단어들은 모두 모음으로만 이루어져 있으므로 끝말은 A, E, I, O, U 중에 하나이다

`-` 참고로 끝말잇기에 모든 단어를 사용한다는 보장이 없으니 새롭게 끝말을 이을 때마다 정답을 갱신해나가야 한다

```python
def word_chain(strings):
    max_score = 0
    dp = [{} for _ in range(MAX_BIT + 1)]  # dp[b][last_word]는 사용한 단어의 상태가 b이고 끝말이 last_word일 때 얻을 수 있는 점수의 최댓값
    for i, s in enumerate(strings):
        dp[1 << i][s[-1]] = len(s)
        max_score = max(len(s), max_score)
    for b in range(MAX_BIT):
        for last_word, score in dp[b].items():
            for i, s in enumerate(strings):
                if last_word != s[0]:
                    continue
                b_next = b | (1 << i)
                if b_next == b:
                    continue
                score_next = dp[b_next].get(s[-1], 0)
                dp[b_next][s[-1]] = max(score + len(s), score_next)
                max_score = max(dp[b_next][s[-1]], max_score)
    return max_score


def solution():
    global MAX_BIT
    N = int(input())
    strings = [input() for _ in range(N)]
    MAX_BIT = 2**N - 1
    answer = word_chain(strings)
    print(answer)


solution()

# input
# 3
# AEIOU
# UIU
# EO
```

## 그림 교환

- 문제 출처: [백준 1029번](https://www.acmicpc.net/problem/1029)

`-` 그림 교환 - 비트필드를 활용한 다이나믹 프로그래밍

`-` 그림을 소유했던 사람의 수의 최댓값을 계산하자

`-` 그림을 팔 때 그림을 산 가격보다 크거나 같은 가격으로 팔아야 한다

`-` 같은 그림을 두 번 이상 사는 것은 불가능하다

`-` 1번 아티스트는 그 그림을 외부 상인에게 가격 $0$을 주고 샀다

`-` 그림을 구매한 아티스트의 상태를 비트마스크로 관리하자

`-` 현재 그림을 소유하고 있는 아티스트를 알아야 한다

`-` 아직 그림을 구매하지 않은 아티스트에게 판매가 가능하다면 그림을 판매한다

`-` 모든 비트마스크의 상태에 대해 수행하면 된다

`-` 위의 과정 중 여태까지의 소유자 수의 최댓값을 초과한다면 최댓값을 갱신한다

`-` 비트마스크 상태 공간은 $O\left(2^N\right)$, 그림 소유자 판단은 $O(N)$, 그림을 팔아넘길 아티스트 순회는 $O(N)$이다

`-` 따라서 전체 알고리즘의 시간복잡도는 $O\left(N^2 2^N\right)$이다 (그림 가격이 모두 $0$인 최악의 경우)

`-` 그림 가격이 모두 같으면 `dp[b][owner_now] == INF`에서 걸러지지 않는다

```python
def bitmask_dp(artist2purchase_prices):
    max_bit = 2**N - 1
    answer = 1
    dp = [[INF] * N for _ in range(max_bit + 1)]  # dp[b][n]은 상태 공간 b, 현재 소유자가 n번 아티스트일 때 구매 가격의 최솟값
    dp[1 << 0][0] = 0  # 소유주가 구매한 가격 (1번 아티스트가 상인에게 가격 0에 구매)
    for b in range(1, max_bit):
        n = bin(b).count("1")
        for owner_now in range(N):
            if dp[b][owner_now] == INF:
                continue
            for owner_next in range(N):
                b_next = b | (1 << owner_next)
                if b == b_next:
                    continue
                if dp[b][owner_now] > artist2purchase_prices[owner_now][owner_next]:
                    continue
                dp[b_next][owner_next] = min(artist2purchase_prices[owner_now][owner_next], dp[b_next][owner_next])
                answer = max(n + 1, answer)
    return answer


def solution():
    global N, INF
    N = int(input())
    artist2purchase_prices = [list(map(int, list(input()))) for _ in range(N)]
    INF = float("inf")
    answer = bitmask_dp(artist2purchase_prices)
    print(answer)


solution()

# input
# 3
# 022
# 101
# 110
```

## 컨닝

- 문제 출처: [백준 1014번](https://www.acmicpc.net/problem/1014)

`-` 격자에서 비트마스크 동적 계획법을 사용하면 될 것 같다

`-` 해당 유형을 처음 풀어봐서 어떻게 해야 깔끔하게 계산할 수 있을지 모르겠지만 일단 내 생각대로 해보겠다

`-` 맨 아래 오른쪽 자리부터 시작해 한 줄을 채우며 위쪽 줄로 올라갈 것이다

`-` 각 자리에는 학생이 있거나 없으므로 $2$개의 상태를 가진다

`-` 만약 책상이 부서져 있다면 학생은 해당 자리에 앉을 수 없다

`-` 한 줄은 $M$칸이므로 상태 공간을 이진수로 나타내면 $0$부터 $2^M-1$의 값을 가진다

`-` 자신의 왼쪽, 오른쪽, 왼쪽 대각선 위, 오른쪽 대각선 위에는 학생이 없어야 하므로 실제론 이보다 적은 상태 공간을 가질 것이다

`-` $\operatorname{dp}[i]$를 가장 오른쪽 자리부터 $i$개의 자리를 고려했을 때 가질 수 있는 상태 공간의 모음과 그 때의 학생 수의 최댓값이라 하자

`-` 맨 아래 줄의 경우 이진수로 나태낸 상태 공간에서 $1$의 개수가 자리에 배치된 학생의 수이다

`-` $\operatorname{dp}[i]$는 $\operatorname{dp}[i-1]$을 순회하며 $i$번째 자리에 학생을 배치하거나 배치하지 않을 것이다

`-` 이진수에서 앞자리가 $1$이면 $i$번째 자리에 학생을 배치할 수 없고 $0$이면 배치하거나 배치하지 않는다 (참고로 $\operatorname{dp}[i]$ 계산할 때 $\operatorname{dp}[i-1]$만 알고 있으면 된다)

`-` 이렇게 해서 $O\left(2^M\right)$의 시간 복잡도로 맨 아래 줄에 배치할 수 있는 모든 경우의 수를 구했다

`-` 이를 바탕으로 바로 위의 줄의 배치를 계산할 것이다

`-` 자신의 왼쪽 대각선 위, 오른쪽 대각선 위에는 학생이 없어야 하므로 위의 줄의 배치를 계산할 때 아래 줄의 상태 공간을 필요로 한다

`-` 아래 줄의 상태 공간 각각에 대해 위의 줄이 가질 수 있는 상태 공간을 기록할 것이다

`-` 맨 아래 줄에서 했던 것처럼 하되 아래 줄의 $i$번째 자리에 학생이 있으면 위의 줄의 $i+1$번째와 $i-1$번째 자리에 학생이 없어야 하는 것만 추가하면 된다

`-` 현재 고려하고 있는 아래 줄의 상태 공간을 $b_{\operatorname{lower}}$라고 하자

`-` $\operatorname{dp}[i]$는 $(\text{상태 공간}, \text{배치된 최대 학생 수})$들의 모임이다

`-` $i$번째 자리에 학생을 배치할려면 $i-1$번째 자리에 학생이 없어야 하고 아래 줄의 $i-1,i+1$번째 자리에도 학생이 없어야 한다

`-` 위의 줄이 완성됐다고 해보자

`-` 이제 또 위의 줄의 상태 공간을 계산해야 되는데 위의 줄의 상태 공간을 계산하는 건 바로 아래 줄의 상태 공간만 필요로 한다

`-` 즉, 처음에 사용한 맨 아래 줄의 상태 공간은 필요없다

`-` $b_{\operatorname{lower}}$를 바탕으로 완성시킨 상태 공간 $b$를 생각해보자

`-` $b$는 꼭 $b_{\operatorname{lower}}$가 아니어도 $b_{\operatorname{lower_2}}$로부터 파생될 수도 있다

`-` 어차피 다음 위의 줄을 계산하는 건 $b$만 필요로 하니 배치된 학생 수를 최대로 해야 이득이다

`-` 따라서 동일한 $b$에 대해 배치된 학생 수를 이전보다 크게 갱신할 수 있다면 최댓값으로 갱신하자

`-` 이러한 작업을 맨 위의 줄까지 반복하면 된다

`-` 맨 아래 줄을 완성하는 건 $O\left(2^M\right)$이고 이를 바탕으로 위의 줄의 상태 공간을 계산하는 건 $O\left(2^{2M}\right)$이다

`-` 총 $N$줄 존재하므로 전체 알고리즘의 시간 복잡도는 $O\left(N 2^{2M}\right)$이다

`-` 물론 각 줄마다 $2^M$개의 상태 공간을 가질 수 있는 건 아니라 시간 복잡도에 비해 실제 연산은 적을 것이다

`-` $N,M \le 10 $이므로 제한 시간 안에 작동할 수 있다

```python
from collections import defaultdict


def bottom_up(classroom):
    lower = [(0, 0)]  # base
    for x in range(N):
        bin2max = defaultdict(int)
        for b_lower, n_lower in lower:
            prev = [(0, n_lower)]
            if classroom[N - x - 1][M - 1] != IMPOSSIBLE and b_lower & (1 << 1) == 0:
                prev.append((1, n_lower + 1))
            for i in range(1, M):
                if classroom[N - x - 1][M - i - 1] == IMPOSSIBLE:
                    continue
                now = []
                for b, n in prev:
                    if b & (1 << (i - 1)) or b_lower & (1 << (i - 1)):
                        continue
                    if i < M - 1 and b_lower & (1 << (i + 1)):
                        continue
                    b_new = b | (1 << i)
                    now.append((b_new, n + 1))
                now.extend(prev)
                prev = now
            for b, n in prev:
                bin2max[b] = max(n, bin2max[b])
        lower = list(bin2max.items())
    max_students = 0
    for _, n in lower:
        max_students = max(n, max_students)
    return max_students


def solve_testcase():
    global N, M, IMPOSSIBLE
    N, M = map(int, input().split())
    classroom = [input() for _ in range(N)]
    IMPOSSIBLE = "x"
    answer = bottom_up(classroom)
    print(answer)


def solution():
    C = int(input())
    for _ in range(C):
        solve_testcase()


solution()

# input
# 1
# 2 3
# x.x
# x.x
```

## 불 끄기

- 문제 출처: [백준 14939번](https://www.acmicpc.net/problem/14939)

`-` 격자판에서 비트마스크 동적 계획법을 사용하는 문제이다

`-` 해당 유형을 접해봤으면 이 문제도 충분히 해결할 수 있다

`-` 전구에 달린 스위치를 누르면 그 전구와 위, 아래, 왼쪽, 오른쪽에 있는 전구의 상태도 바뀐다

`-` 이는 임의의 라인에 영향을 끼치는 건 그 위의 라인과 아래 라인뿐이란 걸 시사한다

`-` 전구 $100$개가 $10\times 10$ 정사각형 모양으로 늘어서 있다

`-` 전구는 꺼져있거나 켜져있다

`-` $1$행부터 $i-1$행까지 모든 전구를 껐다고 했을 때 가능한 $i$행의 상태를 고려해보자

`-` 기본적으로 각 전구는 $2$개의 상태를 가지므로 $i$행의 상태 공간은 $2^{10}$의 크기를 가진다

`-` $i$행 위로는 모든 전구를 껐으니 $i$행의 각 상태 공간에 대해 $i+1$행을 조작하여 $i$행의 전구를 모두 꺼보자

`-` $i+1$행도 $2^{10}$의 크기의 상태 공간을 가진다

`-` 이때 $i$행의 상태 공간과 $i+1$행의 상태 공간이 서로 모순되지 않아야 한다

`-` 이를 확인하기 위해 이전 행의 전구 상태와 이전 행에서 어떤 버튼을 눌렀는지를 알아야 한다

`-` 이전 행의 전구 상태와 이전 행에서 어떤 버튼을 눌렀는지를 바탕으로 이전 행의 전구 상태를 갱신한다

`-` 이번 행에서 어떤 버튼을 누를지 결정하자

`-` 이는 $2^{10}$의 경우의 수를 가진다

`-` 각 경우의 수에 따라 이전 행의 전구 상태를 갱신시키자

`-` 만약 이전 행의 전구를 모두 끄지 못한다면 모든 전구를 끌 수 없으므로 실패이다

`-` 즉, 해당 경우의 수는 불가능한 경우의 수이므로 제외해야 한다

`-` 이를 모든 행에 대해 반복한 뒤 모든 전구가 꺼져 있으면 성공이다

`-` 스위치를 최소한으로 눌러야 되므로 각 상태 공간에 대해 최솟값만 저장하자

`-` 격자판을 $N$행 $N$열로 나타낼 때 알고리즘의 시간 복잡도는 $O\left(N^2 2^{2N}\right)$이다

`-` 비트 연산은 검색해서 조금 찾아봤다

```python
def press_buttons(state, how, n):
    for i in range(n):
        if how & (1 << i):
            # 스위치 온
            state ^= (1 << i)
            if i > 0:
                state ^= (1 << (i - 1))
            if i < n - 1:
                state ^= (1 << (i + 1))
    return state


def press_adj_row_buttons(state, how):
    # 좌우 버튼에는 영향을 주지 않음
    return state ^ how


def bit_count(num):
    return bin(num)[2:].count("1")


def check_all_off(state):
    return state == 0


def bitmask_dp(matrix, n):
    max_bit = 2**n
    # base case (1행)
    dp = {}  # dp[(이전 행 상태, 이전 행에서 뭘 눌렀는지)] = 누른 스위치 개수의 최솟값 
    for how in range(max_bit):
        dp[matrix[0], how] = bit_count(how)
    # 2행부터 n행까지
    for i in range(1, n):
        temp = {}
        for (state_prev_before, how_prev), count_prev in dp.items():
            state_prev_after = press_buttons(state_prev_before, how_prev, n)  # 이전 행에서 버튼을 누른대로 눌러 상태를 변경
            for how_now in range(max_bit):
                state_prev_updated = press_adj_row_buttons(state_prev_after, how_now)  # 이번 행에서 누를 걸 이전 행에 적용시킴, 0이어야 함
                state_now = press_adj_row_buttons(matrix[i], how_prev)
                count_now = bit_count(how_now)
                if not check_all_off(state_prev_updated):
                    continue
                temp[state_now, how_now] = min(count_prev + count_now, temp.get((state_now, how_now), INF))
        dp = temp
    answer = INF
    for (state, how), count in dp.items():
        state = press_buttons(state, how, n)
        if not check_all_off(state):
            continue
        answer = min(count, answer)
    return answer


def solution():
    global INF
    n = 10
    on = "O"
    off = "#"
    INF = float("inf")
    matrix = [int("".join("1" if x == on else "0" for x in input()), 2) for _ in range(n)]
    answer = bitmask_dp(matrix, n)
    if answer == INF:
        print(-1)
        return
    print(answer)


solution()

# input
# #O########
# OOO#######
# #O########
# ####OO####
# ###O##O###
# ####OO####
# ##########
# ########O#
# #######OOO
# ########O#
```

`-` 그리디 알고리즘으로 풀 수 있었다 (비트마스크 동적 계획법에 매몰돼서 생각 못했다)

`-` 첫 번째 줄에 대해 스위치를 누르는 경우의 수는 $2^{10}$이다

`-` 왜냐하면 같은 스위치를 $2$번 누르는 것은 $0$번 누르는 것과 마찬가지이므로 그 이상 누를 필요가 없다

`-` 따라서 스위치를 누르거나 안 누르거나 $2$가지이고 한 줄에 스위치가 $10$개이므로 $2^{10}$이다

`-` 각 경우에 따라 스위치를 눌렀다고 하자

`-` 그럼 첫 번째 줄에서 아직 켜져있는 스위치를 끄기 위해 바로 아래의 스위치를 눌러야 한다

`-` 즉, 해야 되는 행동이 정해져 있다

`-` 이를 마지막 줄까지 수행한 뒤 마지막 줄의 스위치가 모두 꺼져있으면 모든 스위치를 끌 수 있는 것이고 여태까지 누른 스위치의 개수가 정답의 후보가 된다

`-` 정답의 후보 중 최솟값이 정답이 된다

`-` 알고리즘의 시간 복잡도는 $O\left(N^2 2^N\right)$이다

## 몬드리안의 꿈

- 문제 출처: [백준 6569번](https://www.acmicpc.net/problem/6569)

`-` 격자에서 비트마스크 DP를 계산하는 문제이다

`-` [컨닝](https://www.acmicpc.net/problem/1014) 문제를 먼저 풀고 오는 것이 좋다

`-` [격자판 채우기](https://www.acmicpc.net/problem/1648) 문제를 풀기 전에 먼저 풀려고 한다

`-` 큰 직사각형은 격자로 작은 직사각형은 도미노로 생각하겠다

`-` 큰 직사각형의 높이를 $N$, 너비를 $M$이라 하자

`-` $2 \times 1$ 크기의 도미노는 $0$으로, $1 \times 2$ 크기의 도미노는 $1$로 표현하자

`-` 그럼 하나의 행은 $0$과 $1$로 구성된 이진법으로 표시된다

`-` 만약 $0$이 연속하여 홀수 개만 존재하면 모순이므로 불가능한 상태 공간이다

`-` 가로 크기가 $M$이므로 $M$개의 비트를 갖는 $2$진수 중 가능한 것만 미리 계산해두자

`-` 근데 $1 \times 2$ 크기의 도미노의 경우 위 칸이냐 아래 칸이냐에 따라 다르게 취급해야 한다

`-` 어떤 열의 위의 행이 $1 \times 2$ 크기 도미노의 위 칸이면 자신은 무조건 아래 칸이어야 한다 

`-` 따라서 비트가 $1$일 때 이것이 위 칸인지 아래 칸인지 나타내는 제약 조건의 상태 공간도 이진법으로 정의하겠다

`-` 그럼 어떤 행의 상태 공간은 두 개의 $2$진수로 표현할 수 있다

`-` 격자판을 도미노로 채울 때 바로 직전의 행만 배치에 영향을 주기 때문에 직전의 상태 공간만 저장하고 있자

`-` 직전의 상태 공간을 바탕으로 현재 올 수 있는 상태 공간을 정의하고 각 상태 공간별로 격자판을 채우는 방법의 수를 가지고 있자

`-` 직전의 상태 공간에서 격자판을 채우는 방법의 수를 현재 상태 공간에서 격자판을 채우는 방법의 수에 누적하면 된다

`-` 이를 마지막 행까지 배치가 모순되지 않게 반복한 뒤 모든 상태 공간 별로 격자판을 채우는 방법의 수를 더해주면 정답이 된다

`-` 마지막 행까지 진행했으면 제약 조건의 상태 공간은 $0$이어야 한다

`-` 참고로 $N\times M$이 홀수이면 격자판을 빈 공간이 없게 채울 수 없어서 정답은 $0$이다

```python
from collections import defaultdict


def check_possible_width_domino_batch(batch, col_size):
    j = 0
    for i in range(col_size):
        if (batch & (1 << i)):
            if j % 2 == 1:
                return False
            j = 0
            continue
        j += 1
    return j % 2 == 0


def check_possible_height_domino_batch(batch, condition, col_size):
    condition_new = 0
    for i in range(col_size):
        if condition & (1 << i) == 0:
            if batch & (1 << i) == 0:
                continue 
            condition_new |= (1 << i)
        if batch & (1 << i) == 0:
            return IMPOSSIBLE
    return condition_new


def count_tilings(row_size, col_size, possible_batches):
    dp = {}  # 상태 공간(0은 가로, 1은 세로), 세로 도미노의 위 칸 여부
    for batch in possible_batches:
        dp[batch, batch] = 1  # 초깃값
    for _ in range(row_size - 1):
        temp = defaultdict(int)
        for (batch_prev, condition_prev), count in dp.items():
            for batch_now in possible_batches:
                condition_now = check_possible_height_domino_batch(batch_now, condition_prev, col_size)
                if condition_now == IMPOSSIBLE:
                    continue
                temp[batch_now, condition_now] += dp[batch_prev, condition_prev]
        dp = temp
    return sum(value for (_, condition), value in dp.items() if condition == 0)


def solve_testcase(h, w):
    if (h * w) % 2 == 1:
        print(0)
        return
    possible_batches = [batch for batch in range(2**w) if check_possible_width_domino_batch(batch, w)]
    answer = count_tilings(h, w, possible_batches)
    print(answer)


def solution():
    global IMPOSSIBLE
    IMPOSSIBLE = -1
    while True:
        h, w = map(int, input().split())
        if h == 0 and w == 0:
            break
        solve_testcase(h, w)


solution()

# input
# 1 2
# 1 3
# 1 4
# 2 2
# 2 3
# 2 4
# 2 11
# 4 11
# 0 0
```
