# 깊이 우선 탐색 (Depth-First Search)

`-` 시작 노드에서 출발해 갈 수 있는 데까지 깊이 탐색한 뒤 막히면 이전 분기로 되돌아가서 다른 경로를 탐색

## ACM Craft

- 문제 출처: [백준 1005번](https://www.acmicpc.net/problem/1005)

`-` 사이클 없는 단방향 그래프이며 목표 건물 $W$에서 역방향으로 거슬러 올라가는 걸 생각해보자

`-` 다음 건물을 짓기 위해선 해당 건물 이전에 지어야 하는 건물을 모두 지어야 한다

`-` 이전 건물들 중 가장 오래 걸리는 시간이 다음 건물을 짓기 위한 준비 시간이다

`-` 임의의 건물을 $u$라 하고 $u$를 짓기 위해 건설해야 하는 건물들을 $p_1,p_2,\cdots p_n$이라고 하자

`-` 어떤 건물 $u$을 짓기 위해 소요되는 시간을 $W(u)$라고 하고 건물 $u$를 건설하기 시작하여 종료하는데 걸리는 시간을 $T_u$라고 하자

`-` 그럼 $W(u) = T_u + \max\left(W(p_1), W(p_2), \cdots ,W(p_n) \right)$이다

`-` 만약 미리 지어야 하는 건물이 없다면 $W(u) = T_u$이다

`-` 이를 목표 건설 $W$에서 시작하여 재귀적으로 적용하자

`-` 이미 탐색한 건물을 또 방문할 수도 있어서 메모이제이션을 해야 한다 

`-` 역방향으로 움직이기 위해 노드 클래스를 만들어 부모 노드(이전 건물)를 저장하자

`-` 처음엔 $W$와 연결된 그래프만 추출하고 루트에서 이동하는 걸 고려했는데 생각할게 많아 $W$부터 거슬러 올라가는 방법으로 바꿨다

```python
class Node:
    def __init__(self, key):
        self.key = key
        self.prev = []

    def add_prev(self, node):
        self.prev.append(node)


def dfs(node):
    if node.key in dp:
        return dp[node.key]
    max_time = 0
    for prev in node.prev:
        time = dfs(prev)
        max_time = max(time, max_time)
    max_time += D_n[node.key]
    dp[node.key] = max_time
    return max_time


def solve_testcase():
    global D_n, dp
    N, K = map(int, input().split())
    D_n = [0] + list(map(int, input().split()))  # 건물 번호는 1부터 시작
    num2node = {i: Node(i) for i in range(1, N + 1)}
    dp = {}  # dp[i]는 건물 i를 짓기 위해 소요되는 시간
    for _ in range(K):
        prev, now = map(int, input().split())
        num2node[now].add_prev(num2node[prev])
    W = int(input())
    answer = dfs(num2node[W])
    print(answer)


def solution():
    T = int(input())
    for _ in range(T):
        solve_testcase()


solution()

# input
# 2
# 4 4
# 10 1 100 10
# 1 2
# 1 3
# 2 4
# 3 4
# 4
# 8 8
# 10 20 1 5 8 7 1 43
# 1 2
# 1 3
# 2 4
# 2 5
# 3 6
# 5 7
# 6 7
# 7 8
# 7
```

## 텀 프로젝트

- 문제 출처: [백준 9466번](https://www.acmicpc.net/problem/9466)

`-` 임의의 번호 $x$에서 시작하여 해당 학생이 가리키는 번호로 이동하는 것을 반복했을 때 $x$로 끝나야 해당 사이클에 속한 학생들이 프로젝트를 수행할 수 있다

`-` 즉, 사이클에 속하지 못한 학생은 프로젝트 팀에 속하지 못한다

`-` 모든 학생은 누군가를 가리키므로 사이클이 적어도 하나 존재한다

`-` 임의의 번호에서 시작해 가리키는 번호를 따라가다가 이미 방문한 번호를 $2$번 방문한다면 사이클이 존재하는 것이다

`-` 해당 번호에서 이전에 방문한 번호는 사이클이 없고 이후에 방문한 번호는 사이클이 있다

`-` 해당 사이클 집합에 속한 사람들은 프로젝트 팀에 속한 것이며 해당 집합에 속하지 않은 학생이 사이클 집합에 속한 번호를 가리키면 해당 학생은 사이클에 속하지 못한다

```python
import sys

sys.setrecursionlimit(10**5)


def dfs(node, nth, graph, visited):
    if node in in_team or node in no_team:
        visited[node] = INF
        return node, visited
    if node in visited:
        return node, visited  
    visited[node] = nth
    next_ = graph[node]
    return dfs(next_, nth + 1, graph, visited)


def solve_testcase():
    global INF, in_team, no_team
    n = int(input())
    choices = [0] + list(map(int, input().split()))
    INF = float("inf")
    in_team = set()
    no_team = set()
    for u in range(1, n + 1):
        if u in in_team or u in no_team:
            continue
        visited = {}
        node, visited = dfs(u, 1, choices, visited)
        nth = visited[node]
        for v, i in visited.items():
            if i < nth:
                no_team.add(v)
                continue
            in_team.add(v)
    answer = len(no_team)
    print(answer)


def solution():
    T = int(input())
    for _ in range(T):
        solve_testcase()


solution()

# input
# 2
# 7
# 3 1 3 7 3 4 6
# 8
# 1 2 3 4 5 6 7 8
```

## 임계경로

- 문제 출처: [백준 1948번](https://www.acmicpc.net/problem/1948)

`-` 그래프에 사이클이 없으므로 **한걸음**에서 **로마**까지 가는데 같은 도시는 한 번만 방문할 수 있다

`-` 또한 **한걸음**에선 어떤 도시로든 갈 수 있고 어떤 도시든 **로마**까지 갈 수 있다

`-` 일단 SPFA 알고리즘을 통해 **한걸음**에서 **로마**까지 가는 최장 시간 경로를 찾을 수 있다 (간선을 음수로 바꾼 뒤 수행)

`-` 출발 도시인 **한걸음**을 $s$, 도착 도시인 **로마**를 $t$로 나타내자

`-` SPFA 알고리즘으로 찾아낸 선택받은 사람들이 모두 **로마**에 도착하는데 걸리는 시간을 $T$라고 하자

`-` $u$에서 $t$로 가는데 각각 $a, b$의 시간이 걸리는 경로가 존재한다고 해보자 ($ a<b$)

`-` 그럼 $s$에서 $u$까지 가는데 $T-a, T-b$ 시간에 도착할 수 있다면 $t$까지 $T$의 시간에 갈 수 있다는 것이다

`-` 그런데 $s$에서 $u$까지 $T-a$의 시간이 걸리는 경로로 간 다음 $u$에서 $t$까지 가는데 $b$의 시간이 걸리는 경로를 이용하면 총 시간이 $T$보다 크다

`-` 이는 처음에 가정했던 것과 모순이므로 어떤 도시에서 **로마**까지 가는데 걸리는 시간은 유일해야 하며 이는 최대 시간이어야 한다

`-` 가장 오래 달린 사람들이 달린 도로에 전부 황금을 칠할 것이므로 $s$에서 $t$까지 $T$ 시간에 도착하는 경로에 포함된 도로 개수를 추적할 수 있어야 한다

`-` $\operatorname{dp}[u]$를 $s$에서 $u$로 가는데 소요되는 최장 시간이라고 하자

`-` 그럼 단방향 간선으로 연결된 $r_i \to u$인 $r_1, r_2, \dots, r_n$에 대해 $\operatorname{dp}[u] = \max(\operatorname{dp}[r_1] + w_1, \operatorname{dp}[r_2] + w_2, \dots ,\operatorname{dp}[r_n] + w_n)$이다 (단, $w_i$는 $r_i$에서 $u$로 가는데 걸리는 시간)

`-` 초깃값으로 $\operatorname{dp}[s] = 0$이다

`-` 어떤 도로 $u\to v$에 대해 해당 도로를 황금으로 칠할려면 조건 $1$로 $\operatorname{dp}[v]$와 $\operatorname{dp}[u] + w_{uv}$가 같아야 하고 조건 $2$로 $v$에서 $t$로 $\operatorname{dp}[t]-\operatorname{dp}[v]$ 시간에 도착할 수 있어야 한다

`-` 역추적은 **로마**에서 시작하자

`-` $t$와 연결된 $v_i$에 대해 $\operatorname{dp}[t]$와 $\operatorname{dp}[v_i] + w_{v_it}$가 같은지 확인하자

`-` 이것이 동일하다면 $v_i \to t$ 도로를 황금으로 칠할 수 있다 ($\operatorname{dp}[t] - \operatorname{dp}[t] = 0$이다)

`-` 그리고 이를 만족하는 $v_i$에 대해 재귀적으로 확인하자 (이미 방문한 도시는 다시 방문하지 않아도 된다)

`-` 당연하게 $v_i$에서 $t$로 $\operatorname{dp}[t]-\operatorname{dp}[v_i]$ 시간에 갈 수 있으므로 재귀적으로 확인하는 도시에 대해 조건 $1$만 충족하면 된다

`-` 참고로 $\operatorname{dp}[t]$가 $T$이므로 SPFA 알고리즘을 사용하지 않아도 된다

`-` 재귀적으로 모든 노드와 간선을 최대 한 번만 탐색하므로 $\operatorname{dp}[t]$를 찾는 것과 역추적 알고리즘의 시간 복잡도는 $O(V+E)$이다 ($V$는 노드의 개수, $E$는 간선의 개수)

`-` [ACM Craft](https://www.acmicpc.net/problem/1005) 문제에 역추적을 더한 문제였다

```python
import sys

sys.setrecursionlimit(10**4 + 2)


def dfs(v, adj_list, weight, dp):
    if dp[v] > -INF:
        return dp[v]
    for u in adj_list[v]:
        if dfs(u, adj_list, weight, dp) + weight[u, v] <= dp[v]:
            continue
        dp[v] = dp[u] + weight[u, v]
    return dp[v]


def track(v, adj_list, weight, dp, visited):
    global COUNT
    visited[v] = True
    for u in adj_list[v]:
        if dp[u] + weight[u, v] < dp[v]:
            continue
        COUNT += 1
        if visited[u]:
            continue
        track(u, adj_list, weight, dp, visited)


def solution():
    global COUNT, INF
    n = int(input())
    m = int(input())
    adj_list = [[] for _ in range(n + 1)]
    weight = {}
    for _ in range(m):
        u, v, w = map(int, input().split())
        adj_list[v].append(u)
        weight[u, v] = w
    start, end = map(int, input().split())
    INF = float("inf")
    dp = [-INF] * (n + 1)
    dp[start] = 0
    arrival_time = dfs(end, adj_list, weight, dp)
    COUNT = 0
    visited = [False] * (n + 1)
    track(end, adj_list, weight, dp, visited)
    print(arrival_time)
    print(COUNT)


solution()

# input
# 7
# 9
# 1 2 4
# 1 3 2
# 1 4 3
# 2 6 3
# 2 7 5
# 3 5 1
# 4 6 4
# 5 6 2
# 6 7 5
# 1 7
```
