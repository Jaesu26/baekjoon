# 최대 부분 배열 문제 (Maximum Subarray)

## 연속합

- 문제 출처: [백준 1912번](https://www.acmicpc.net/problem/1912)

`-` 규칙을 떠올리기 은근 어려웠다

`-` 위치상 연속된 원소들에 초점을 맞춰 생각하다가 $n$번째 원소를 마지막으로 가지는 연속합에 대해 생각하여 규칙을 떠올렸다

`-` $\text{dp[$n$]}$을 마지막 원소로 $n$번째 원소를 가지는 연속합중 최댓값이라고 하자

`-` 그러면 $\text{dp[$n$]} = \max(a_n,\,a_n+\text{dp[$n-1$]})$이 성립한다 $\cdots (1)$

`-` 왜?

`-` 마지막 원소로 $n$번째 원소를 가지는 연속합으로 가능한 경우는 아래와 같다

`-` $[a_n[1]+\cdots+a_n[n]],\,[a_n[2]+ \cdots + a_n[n]],\, \cdots,\, [a_n[n-1]+a_n[n]],\, [a_n[n]]$

`-` 위의 연속합을 첫 번째부터 $b_1, b_2,\cdots,b_n$이라고 하자

`-` 이들은 모두 $a_n[n]$을 포함하고있어 각각의 연속합에 대해 $a_n[n]$을 차감하더라도 이들의 대소관계는 변하지 않는다

`-` 그러므로 식 $(1)$이 자연스럽게 성립한다

`-` 즉 $n$번째 원소를 마지막으로 가지는 연속합중 최댓값을 얻기위해선 $n-1$번째 원소를 마지막으로 가지는 연속합 중에서 최댓값을 사용해야 한다

`-` 아래의 수열에서 연속합중 최댓값을 구해보자

`-` $10, -4, 3, -35, 21, -1$

`-` $\text{dp[$1$]} = 10$ ---> 당연한 결과이다

`-` $\text{dp[$2$]} = \max(-4, -4 + \text{dp[$1$]}) = 6$ 

`-` $\text{dp[$3$]} = \max(3, 3 + \text{dp[$2$]}) = 9$ 

`-` $\text{dp[$4$]} = \max(-35, -35 + \text{dp[$3$]}) = -26$ 

`-` $\text{dp[$5$]} = \max(21, 21 + \text{dp[$4$]}) = 21$

`-` $\text{dp[$6$]} = \max(-1, -1 + \text{dp[$5$]}) = 20$ 

`-` $\text{dp[$n$]}$의 최댓값은 $21$이다 

```python
n = int(input())
a_n = list(map(int, input().split()))
dp = [0] * n
dp[0] = a_n[0]
for i in range(1, n):
    dp[i] = max(a_n[i], a_n[i] + dp[i - 1])
print(max(dp))

# input
# 10
# 10 -4 3 1 5 6 -35 12 21 -1
```

## 연속합과 쿼리

- 문제 출처: [백준 16993번](https://www.acmicpc.net/problem/16993)

`-` 동적 계획법 단계의 그 문제를 기억하시나요? 이제 그 문제를 $10$만 번 풀 수 있습니다!

`-` 우와아와와와와와와와아아아와와와와아아왕

`-` 세그먼트 트리를 사용해 연속합을 $10$만 번 계산해보자

`-` 세그먼트 트리상에서 노드는 자신이 가리키는 구간 내의 가장 큰 연속합을 저장하고 있어야 한다

`-` 어떤 노드가 구간 $[s,e]$에서 가장 큰 연속합을 저장하고 있다고 하자

`-` 그런데 이를 어떻게 계산할 수 있을까?

`-` 부모 노드의 왼쪽 자식은 구간 $[s, m]$을 담당하며 오른쪽 자식은 구간 $[m + 1, e]$를 담당한다 (단, $m = \left\lfloor\frac{s+e}{2}\right\rfloor$)

`-` 부모 노드에 저장된 연속합을 $C$, 왼쪽 자식에 저장된 연속합을 $C_l$, 오른쪽 자식에 저장된 연속합을 $C_r$이라 하자

`-` 단순히 $C$는 $\max(C_l, C_r)$일 수 있다

`-` 그게 아니라면 $C$는 구간 $[m, m+1]$을 연속합에 포함할 것이다

`-` 그게 아니라는 건 $C$를 구간 $[s,m]$ 또는 $[m+1, e]$에서만 계산해서 얻었다는 것이다

`-` 그런데 구간 $[s,m]$에서 가장 큰 연속합은 $C_l$이고 구간 $[m+1, e]$에서 가장 큰 연속합은 $C_r$이다

`-` 따라서 $C$가 $\max(C_l, C_r)$가 아니라면 $C$를 도출한 구간은 $[m, m+1]$을 포함해야 한다

`-` 구간 $[m, m+1]$을 연속합 구간에 포함하는 걸 알아도 문제는 구간을 얼마나 확장할지 모른다는 것이다

`-` 구간 $[s,e]$에서 가장 큰 연속합을 이루는 구간이 $[i, j]$라 하자

`-` 이는 $[i, m]$과 $[m+1,j]$로 나눌 수 있다

`-` 이때 $i$와 $j$가 무엇인지 모르기 때문에 $i$와 $j$를 임의의 값으로 상정해 구간 합을 계산해서 크기를 비교해야 한다

`-` $\operatorname{p}[k]$를 첫 번째 원소부터 $k$번째 원소까지의 누적 합으로 정의하자

`-` 그럼 $A_i + \cdots + A_m$은 $\operatorname{p}[m] - \operatorname{p}[i-1]$로 나타낼 수 있다

`-` 이때 $m$과 $m+1$은 상수라는 것에 주목하자

`-` 그럼 왼쪽 자식은 자기 담당 구간에서 가장 큰 연속합에 더해 $\operatorname{p}[s-1]$부터 $\operatorname{p}[m-1]$ 중 최솟값을 가지고 있자

`-` 해당 최솟값을 $\operatorname{p}[m]$에서 빼면 $[i,m]$의 구간 합을 얻을 수 있다

`-` 한편, $A_{m+1} + \cdots A_j$는 $\operatorname{p}[j] - \operatorname{p}[m]$이 된다

`-` 오른쪽 자식은 $\operatorname{p}[m+1]$부터 $\operatorname{p}[e]$ 중 최댓값을 추가로 가지고 있으면 된다

`-` 왼쪽 자식과 오른쪽 자식을 취합해서 도출한 연속합과 $\max(C_l, C_r)$ 중 더 큰 것이 $C$가 된다

`-` 둘을 취합한 결과는 $\operatorname{p}[j] -\operatorname{p}[i-1]$이다

`-` 가독성을 위해 연속합 세그먼트 트리와 누적 합 세그먼트 트리를 따로 관리하자 (둘은 독립된 연산이다)

`-` 쿼리는 세그먼트 트리를 빌드할 때와는 조금 다르게 진행해야 한다

`-` 쿼리는 구간 $[i, j]$에서 가장 큰 연속합을 출력해야 한다

`-` 구간 $[i,j]$를 담당하는 노드가 $1$개가 아닐 수 있다 (일반적으론 $1$개가 아니다)

`-` 따라서 누적 합을 계산할 때 구간 $[i,j]$를 벗어난 인덱스에선 값을 얻지 않도록 해야 한다

`-` 전체 알고리즘의 시간 복잡도는 $O\left(N\log N+M\log^2 N \right)$이다

`-` 시간 초과가 발생해서 누적 합 쿼리를 메모이제이션 했더니 맞혔다

`-` 구간의 한 쪽 끝은 고정이기에 히트율이 괜찮다

`-` 근데 메모이제이션 안 해도 통과할 줄 알았는데 사용해도 $7068\operatorname{ms}$로 간신히 통과했다

```python
import math


def init_min_max_tree(array, tree, node, start, end):
    if start == end:
        tree[node] = [array[start], array[start]]
        return
    mid = (start + end) // 2
    l = 2 * node
    r = 2 * node + 1
    init_min_max_tree(array, tree, l, start, mid)
    init_min_max_tree(array, tree, r, mid + 1, end)
    min_ = tree[l][MIN] if tree[l][MIN] < tree[r][MIN] else tree[r][MIN]
    max_ = tree[l][MAX] if tree[l][MAX] > tree[r][MAX] else tree[r][MAX]
    tree[node] = [min_, max_]


def min_prefix_sum_query(left, right, tree, node, start, end, dp):
    if left > end or right < start:
        return INF
    if left <= start and end <= right:
        return tree[node][MIN]
    if (left, right) in dp:
        return dp[left, right]
    mid = (start + end) // 2
    left_min = min_prefix_sum_query(left, right, tree, 2 * node, start, mid, dp)
    right_min = min_prefix_sum_query(left, right, tree, 2 * node + 1, mid + 1, end, dp)
    if left_min < right_min:
        return left_min
    return right_min


def max_prefix_sum_query(left, right, tree, node, start, end, dp):
    if left > end or right < start:
        return -INF
    if left <= start and end <= right:
        return tree[node][MAX]
    if (left, right) in dp:
        return dp[left, right]
    mid = (start + end) // 2
    left_max = max_prefix_sum_query(left, right, tree, 2 * node, start, mid, dp)
    right_max = max_prefix_sum_query(left, right, tree, 2 * node + 1, mid + 1, end, dp)
    if right_max > left_max:
        return right_max
    return left_max


def init_max_subarray_sum_tree(array, min_max_prefix_sum_tree, tree, node, start, end, dp_min, dp_max):
    if start == end:
        tree[node] = array[start]
        return
    mid = (start + end) // 2
    init_max_subarray_sum_tree(array, min_max_prefix_sum_tree, tree, 2 * node, start, mid, dp_min, dp_max)
    init_max_subarray_sum_tree(array, min_max_prefix_sum_tree, tree, 2 * node + 1, mid + 1, end, dp_min, dp_max)
    p_i_1 = min_prefix_sum_query(start - 1, mid - 1, min_max_prefix_sum_tree, 1, 0, N, dp_min)
    dp_min[start - 1, mid - 1] = p_i_1
    p_j = max_prefix_sum_query(mid, end, min_max_prefix_sum_tree, 1, 0, N, dp_max)
    dp_max[mid, end] = p_j
    tree[node] = max(tree[2 * node], p_j - p_i_1, tree[2 * node + 1])


def max_subarray_sum_query(left, right, min_max_prefix_sum_tree, tree, node, start, end, dp_min, dp_max):
    if left > end or right < start:
        return -INF
    if left <= start and end <= right:
        return tree[node]
    mid = (start + end) // 2
    left_max_subarray_sum = max_subarray_sum_query(left, right, min_max_prefix_sum_tree, tree, 2 * node, start, mid, dp_min, dp_max)
    right_max_subarray_sum = max_subarray_sum_query(left, right, min_max_prefix_sum_tree, tree, 2 * node + 1, mid + 1, end, dp_min, dp_max)
    if left_max_subarray_sum == -INF:
        return right_max_subarray_sum
    if right_max_subarray_sum == -INF:
        return left_max_subarray_sum
    in_ = start if start > left else left
    out = end if end < right else right
    p_i_1 = min_prefix_sum_query(in_ - 1, mid - 1, min_max_prefix_sum_tree, 1, 0, N, dp_min)
    dp_min[in_ - 1, mid - 1] = p_i_1
    p_j = max_prefix_sum_query(mid, out, min_max_prefix_sum_tree, 1, 0, N, dp_max)
    dp_max[mid, out] = p_j
    return max(left_max_subarray_sum, p_j - p_i_1, right_max_subarray_sum)


def generate_prefix_sums(array):
    prefix_sums = [0] * len(array)
    for i, a_i in enumerate(array[1:], start=1):
        prefix_sums[i] = prefix_sums[i - 1] + a_i
    return prefix_sums


def solution():
    global N, INF, MIN, MAX
    N = int(input())
    array = [0]
    array.extend(map(int, input().split()))
    M = int(input())
    INF = float("inf")
    MIN, MAX = 0, 1
    h = math.ceil(math.log2(N + 1))
    tree_size = 2**(h + 1)
    min_max_prefix_sum_tree = [0] * tree_size
    max_subarray_sum_tree = [0] * tree_size
    start, end = 0, N
    root = 1
    dp_min, dp_max = {}, {}
    prefix_sums = generate_prefix_sums(array)
    init_min_max_tree(prefix_sums, min_max_prefix_sum_tree, root, start, end)
    init_max_subarray_sum_tree(array, min_max_prefix_sum_tree, max_subarray_sum_tree, root, start, end, dp_min, dp_max)
    answers = []
    for _ in range(M):
        i, j = map(int, input().split())
        max_subarray_sum = max_subarray_sum_query(i, j, min_max_prefix_sum_tree, max_subarray_sum_tree, root, start, end, dp_min, dp_max)
        answers.append(max_subarray_sum)
    print("\n".join(map(str, answers)))


solution()

# input
# 10
# 10 -4 3 1 5 6 -35 12 21 -1
# 10
# 1 1
# 3 4
# 1 6
# 2 6
# 6 6
# 7 7
# 8 9
# 8 10
# 1 10
# 5 8
```

`-` 구간 내 누적 합의 최솟값과 최댓값을 호출할 때 하드 코딩을 한 게 아쉽다 ($1,0,N$)

`-` 의미가 명확하기도 하고 함수 인자로 받기엔 이미 인자가 너무 많다

`-` 그리고 야무진 풀이가 있었다 (사람들이 금광 세그로 부른다)

`-` 노드에 구간의 전체 합, 왼쪽부터 시작하는 최대 연속합, 오른쪽부터 시작하는 최대 연속합, 구간 내 최대 연속합을 저장하자

`-` 그리고 왼쪽 자식과 오른쪽 자식을 적절하게 취합해 부모 노드를 갱신하면 된다

`-` 되게 깔끔하고 야무진 풀이 방식이다

## 수열과 쿼리 10

- 문제 출처: [백준 13557번](https://www.acmicpc.net/problem/13557)

`-` [연속합과 쿼리](https://www.acmicpc.net/problem/16993) 문제를 풀 때 사용되는 테크닉이 필요하다

`-` 쿼리가 주어졌을 때 최대 연속합을 구하면 되는데 시작과 끝 인덱스에 제약이 있다

`-` 제약에 더불어 $i \le j$를 만족해야 하니 경우의 수를 나누어 각각에 대해 최대 연속합을 구하자

`-` $y_1 \le x_2$라고 해보자

`-` 그럼 최대 연속합은 적어도 $A_{y_1} + \cdots + A_{x_2}$를 포함한다

`-` 여기에 $[x_2 + 1, y_2]$ 구간의 왼쪽 끝을 포함하는 최대 연속합을 더해야 한다

`-` 그리고 $[x_1, y_1 - 1]$ 구간의 오른쪽 끝을 포함하는 최대 연속합도 더해주면 된다

`-` 만약 음수라면 더하지 말자 (필수적인 부분 합 외에는 양수일 때만 더하자, 한 번 틀리고 깨달음)

`-` 이제 $y_1 > x_2$라고 해보자

`-` 만약 $i < x_2$라면 최대 연속합은 $A_i + \cdots + A_{x_2}$를 포함해야 한다

`-` 이는 $[x_1, x_2]$ 구간의 오른쪽 끝을 포함하는 최대 연속합이다

`-` 여기에 $[x_2 + 1, y_2]$ 구간의 왼쪽 끝을 포함하는 최대 연속합도 더해주면 된다

`-` 만약 $y_1 < j$라면 최대 연속합은 $A_{y_1} + \cdots + A_{j}$를 포함해야 한다

`-` 이는 $[y_1, y_2]$ 구간의 왼쪽 끝을 포함하는 최대 연속합이다

`-` 여기에 $[x_1, y_1 - 1]$ 구간의 오른쪽 끝을 포함하는 최대 연속합도 더해주면 된다

`-` 또는 $x_2 \le i \le j \le y_1$일 수도 있다

`-` 이는 $[x_2, y_1]$ 구간의 최대 연속합이다

`-` 이 중 최댓값이 정답이다

`-` 전체 알고리즘의 시간 복잡도는 $O(N+M\log N)$이다

```python
import math


def init(array, tree, node, start, end):
    if start == end:
        tree[node] = [array[start]] * 4
        return
    mid = (start + end) // 2
    lc = 2 * node
    rc = 2 * node + 1
    init(array, tree, lc, start, mid)
    init(array, tree, rc, mid + 1, end)
    lmax = max(tree[lc][LEFT], tree[lc][SUM] + tree[rc][LEFT])
    rmax = max(tree[rc][RIGHT], tree[lc][RIGHT] + tree[rc][SUM])
    sum_ = tree[lc][SUM] + tree[rc][SUM]
    max_ = max(tree[lc][MAX], tree[lc][RIGHT] + tree[rc][LEFT], tree[rc][MAX])
    tree[node] = lmax, rmax, sum_, max_


def max_left_range_sum_query(left, right, tree, node, start, end):
    if left > right:
        return 0, 0
    if left > end or right < start:
        return -INF, 0
    if left <= start and end <= right:
        return tree[node][LEFT], tree[node][SUM]
    mid = (start + end) // 2
    lc_lmax, lc_sum = max_left_range_sum_query(left, right, tree, 2 * node, start, mid)
    rc_lmax, rc_sum = max_left_range_sum_query(left, right, tree, 2 * node + 1, mid + 1, end)
    return max(lc_lmax, lc_sum + rc_lmax), lc_sum + rc_sum


def max_right_range_sum_query(left, right, tree, node, start, end):
    if left > right:
        return 0, 0
    if left > end or right < start:
        return -INF, 0
    if left <= start and end <= right:
        return tree[node][RIGHT], tree[node][SUM]
    mid = (start + end) // 2
    lc_rmax, lc_sum = max_right_range_sum_query(left, right, tree, 2 * node, start, mid)
    rc_rmax, rc_sum = max_right_range_sum_query(left, right, tree, 2 * node + 1, mid + 1, end)
    return max(lc_rmax + rc_sum, rc_rmax), lc_sum + rc_sum


def range_sum_query(left, right, tree, node, start, end):
    if left > end or right < start:
        return 0
    if left <= start and end <= right:
        return tree[node][SUM]
    mid = (start + end) // 2
    left_sum = range_sum_query(left, right, tree, 2 * node, start, mid)
    right_sum = range_sum_query(left, right, tree, 2 * node + 1, mid + 1, end)
    return left_sum + right_sum


def max_range_sum_query(left, right, tree, node, start, end):
    if left > end or right < start:
        return -INF, -INF, 0, -INF
    if left <= start and end <= right:
        return tree[node]
    mid = (start + end) // 2
    left_child = max_range_sum_query(left, right, tree, 2 * node, start, mid)
    right_child = max_range_sum_query(left, right, tree, 2 * node + 1, mid + 1, end)
    lmax = max(left_child[LEFT], left_child[SUM] + right_child[LEFT])
    rmax = max(left_child[RIGHT] + right_child[SUM], right_child[RIGHT])
    sum_ = left_child[SUM] + right_child[SUM]
    max_ = max(left_child[MAX], left_child[RIGHT] + right_child[LEFT], right_child[MAX])
    return lmax, rmax, sum_, max_


def solution():
    global INF, LEFT, RIGHT, SUM, MAX
    N = int(input())
    array = list(map(int, input().split()))
    M = int(input())
    INF = float("inf")
    LEFT, RIGHT, SUM, MAX = 0, 1, 2, 3
    h = math.ceil(math.log2(N))
    tree_size = 2**(h + 1)
    tree = [0] * tree_size
    start, end = 0, N - 1
    root = 1
    init(array, tree, root, start, end)
    answers = []
    for _ in range(M):
        x1, y1, x2, y2 = map(lambda x: int(x) - 1, input().split())
        if y1 <= x2:
            sum_ = range_sum_query(y1, x2, tree, root, start, end)
            lmax, _ = max_left_range_sum_query(x2 + 1, y2, tree, root, start, end)
            rmax, _ = max_right_range_sum_query(x1, y1 - 1, tree, root, start, end)
            lmax, rmax = max(lmax, 0), max(rmax, 0)
            answer = lmax + sum_ + rmax
        else:
            lmax1, _ = max_left_range_sum_query(x2 + 1, y2, tree, root, start, end)
            rmax1, _ = max_right_range_sum_query(x1, x2, tree, root, start, end)
            lmax2, _ = max_left_range_sum_query(y1, y2, tree, root, start, end)
            rmax2, _ = max_right_range_sum_query(x1, y1 - 1, tree, root, start, end)
            _, _, _, max3 = max_range_sum_query(x2, y1, tree, root, start, end)
            lmax1, rmax2 = max(lmax1, 0), max(rmax2, 0)
            answer = max(lmax1 + rmax1, lmax2 + rmax2, max3)
        answers.append(answer)
    print("\n".join(map(str, answers)))


solution()

# input
# 6
# 3 -2 1 -4 5 2
# 2
# 1 1 2 3
# 1 3 2 5
```

## 바이러스 시뮬레이터

- 문제 출처: [백준 31879번](https://www.acmicpc.net/problem/31879)

`-` 금광 세그라 불리는 테크닉이 재밌어서 관련 문제를 풀어보자 (근데 [금광](https://www.acmicpc.net/problem/10167) 문제는 언제 풀지...)

`-` $3$번 쿼리부터 보자

`-` 쿼리 구간에 대해 전체 그룹 구성원의 면역력 합의 최댓값을 구해야 한다

`-` 이는 구간을 절반으로 나눴을 때 왼쪽 또는 오른쪽 그룹 구성원의 면역력 합의 최댓값일 수 있다

`-` 또는 정답 그룹이 왼쪽과 오른쪽 그룹에 동시에 속할 수도 있다

`-` 그런데 면역력이 있는 사람들만 그룹에 속해야 한다

`-` 왼쪽 그룹의 오른쪽 끝을 포함하는 면역력 그룹 구성원의 면역력 합의 최댓값과 오른쪽 그룹의 왼쪽 끝을 포함하는 면역력 그룹 구성원의 면역력 합의 최댓값을 더한 값이다

`-` 근데 면역력은 $0$ 이상이다 (음수가 없음)

`-` 따라서 그룹에 제약이 없으면 그냥 구간 $[l,r]$의 구간 합이 정답이다

`-` 면역력 합의 최댓값을 구하는 게 목적이니 면역력이 $0$이라면 $-\infty$로 대체하자

`-` $-\infty$을 포함하면 정답이 될 수 없다

`-` 그럼 자연스럽게 면역력이 있는 사람들끼리 만든 그룹만 고려하게 된다

`-` 만약 결과가 $-\infty$이라면 해당 쿼리 구간엔 면역력이 $0$인 사람들만 있는 것이니 $0$이 정답이다

`-` $2$번 쿼리는 값 변경 쿼리이다

`-` 면역력이 $-\infty$인 사람에게 접종하면 면역력이 $x$가 됨에 주의하자

`-` 나머지는 일반적인 값 변경 쿼리와 동일한 방식으로 처리하면 된다

`-` 세그먼트 트리의 노드가 관리하는 값이 특이할 뿐 방식은 다를 게 없다

`-` $1$번 쿼리는 연쇄적인 값 변경 쿼리이다

`-` 바이러스의 생명력이 아무리 커도 모든 사람에게 전이되면 쿼리가 끝난다

`-` 이때 쿼리의 실행 횟수는 $O(N)$이지만 한 번 수행하고 나면 다음부턴 $1$번만 수행된다

`-` 여러 번 수행하도록 할려면 면역력을 올려야 하고 이는 $2$번 쿼리로 수행할 수 있다

`-` 하지만 $2$번 쿼리는 쿼리 $1$번당 $1$명의 면역력만 높일 수 있다

`-` 따라서 모든 $1$번 쿼리에 의해 면역력이 감소되는 사람의 수는 $O(N+Q)$이므로 제한 시간 안에 실행될 수 있다

`-` $1$번 쿼리에서 바이러스의 전이를 구현하기 위해 $\arg\max$를 수행해야 한다

`-` 이를 관리하는 세그먼트 트리를 추가하자

`-` [수열과 쿼리 16](https://www.acmicpc.net/problem/14428) 문제를 풀 때 활용한 세그먼트 트리에서 $\arg\min$을 $\arg\max$로 변경하면 된다

`-` 전체 알고리즘의 시간 복잡도는 $O((N+Q)\log N)$이다

`-` 제한 시간이 $11$초인데 $8976 \operatorname{ms}$로 간신히 통과했다

`-` 근데 실행 시간 변동성이 좀 크다

`-` 동일한 코드인데 실행 시간이 $2$초씩 차이가 난다

```python
import math


def argmax(i, j, array):
    if array[i] > array[j]:
        return i
    if array[j] > array[i]:
        return j
    if i < j:
        return i
    return j


def update_argmax_tree(index, value, array, tree, node, start, end):
    if index < start or index > end:
        return
    if start == end:
        array[index] = value
        tree[node] = index
        return
    mid = (start + end) // 2
    update_argmax_tree(index, value, array, tree, 2 * node, start, mid)
    update_argmax_tree(index, value, array, tree, 2 * node + 1, mid + 1, end)
    tree[node] = argmax(tree[2 * node], tree[2 * node + 1], array)


def range_argmax_query(left, right, array, tree, node, start, end):
    if left > right:
        return NONE
    if left > end or right < start:
        return left
    if left <= start <= end <= right:
        return tree[node]
    mid = (start + end) // 2
    left_argmax = range_argmax_query(left, right, array, tree, 2 * node, start, mid)
    right_argmax = range_argmax_query(left, right, array, tree, 2 * node + 1, mid + 1, end)
    range_argmax = argmax(left_argmax, right_argmax, array)  
    return range_argmax


def update_max_range_sum_tree(index, value, array, tree, node, start, end):
    if index < start or index > end:
        return
    if start == end:
        array[start] = value
        t = tree[node]
        t[LEFT] = t[RIGHT] = t[SUM] = t[MAX] = value
        return
    mid = (start + end) // 2
    lc = 2 * node
    rc = 2 * node + 1
    update_max_range_sum_tree(index, value, array, tree, lc, start, mid)
    update_max_range_sum_tree(index, value, array, tree, rc, mid + 1, end)
    lmax = max(tree[lc][LEFT], tree[lc][SUM] + tree[rc][LEFT])
    rmax = max(tree[rc][RIGHT], tree[lc][RIGHT] + tree[rc][SUM])
    sum_ = tree[lc][SUM] + tree[rc][SUM]
    max_ = max(tree[lc][MAX], tree[lc][RIGHT] + tree[rc][LEFT], tree[rc][MAX])
    tree[node] = lmax, rmax, sum_, max_


def max_range_sum_query(left, right, tree, node, start, end):
    if left > end or right < start:
        return -INF, -INF, 0, -INF
    if left <= start and end <= right:
        return tree[node]
    mid = (start + end) // 2
    left_child = max_range_sum_query(left, right, tree, 2 * node, start, mid)
    right_child = max_range_sum_query(left, right, tree, 2 * node + 1, mid + 1, end)
    lmax = max(left_child[LEFT], left_child[SUM] + right_child[LEFT])
    rmax = max(left_child[RIGHT] + right_child[SUM], right_child[RIGHT])
    sum_ = left_child[SUM] + right_child[SUM]
    max_ = max(left_child[MAX], left_child[RIGHT] + right_child[LEFT], right_child[MAX])
    return lmax, rmax, sum_, max_


def run_query1(i, x, array, argmax_tree, max_range_sum_tree, root, start, end):
    while x > array[i]:
        if array[i] > 0:
            x -= max(array[i], 0)
            update_argmax_tree(i, -INF, array, argmax_tree, root, start, end)
            update_max_range_sum_tree(i, -INF, array, max_range_sum_tree, root, start, end)
        i = range_argmax_query(i + 1, end, array, argmax_tree, root, start, end)
        if i == NONE or array[i] < 0:
            return
    value = array[i] - x if x < array[i] else -INF
    update_argmax_tree(i, value, array, argmax_tree, root, start, end)
    update_max_range_sum_tree(i, value, array, max_range_sum_tree, root, start, end)


def run_query2(i, x, array, argmax_tree, max_range_sum_tree, root, start, end):
    value = max(array[i], 0) + x
    update_argmax_tree(i, value, array, argmax_tree, root, start, end)
    update_max_range_sum_tree(i, value, array, max_range_sum_tree, root, start, end)


def run_query3(l, r, array, argmax_tree, max_range_sum_tree, root, start, end):
    _, _, _, max_range_sum = max_range_sum_query(l, r, max_range_sum_tree, root, start, end)
    max_range_sum = max(max_range_sum, 0)
    return max_range_sum


def solution():
    global INF, NONE, LEFT, RIGHT, SUM, MAX
    N, Q = map(int, input().split())
    array = [0]
    array.extend(map(int, input().split()))
    INF = float("inf")
    NONE = -1
    LEFT, RIGHT, SUM, MAX = 0, 1, 2, 3
    h = math.ceil(math.log2(N + 1))
    tree_size = 2**(h + 1)
    argmax_tree = [0] * tree_size
    max_range_sum_tree = [[-INF] * 4 for _ in range(tree_size)]
    start, end = 1, N
    root = 1
    for i, a in enumerate(array[1:], start=1):
        update_argmax_tree(i, a, array, argmax_tree, root, start, end)
        update_max_range_sum_tree(i, a, array, max_range_sum_tree, root, start, end)
    answers = []
    for _ in range(Q):
        query = list(map(int, input().split()))
        operator = query[0]
        if operator == 1:
            _, i, x = query
            run_query1(i, x, array, argmax_tree, max_range_sum_tree, root, start, end)
        elif operator == 2:
            _, i, x = query
            run_query2(i, x, array, argmax_tree, max_range_sum_tree, root, start, end)
        else:
            _, l, r = query
            max_range_sum = run_query3(l, r, array, argmax_tree, max_range_sum_tree, root, start, end)
            answers.append(max_range_sum)
    print("\n".join(map(str, answers)))


solution()

# input
# 9 8
# 6 3 5 2 3 1 2 3 2
# 1 2 10
# 1 2 5
# 1 3 2
# 3 2 8
# 1 7 2
# 2 4 10
# 1 3 6
# 3 2 9
```
